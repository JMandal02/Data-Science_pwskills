# **Assignment : Clustering**



## Q1 — What is unsupervised learning in the context of machine learning?

**Answer:**  
Unsupervised learning is a family of machine learning methods that operate on data without explicit labels.  
The goal is to discover hidden structures, patterns, or relationships in the input data.

**Key points:**
- No labelled target variable (no `y`); only features `X`.  
- Learns intrinsic structure (clusters, manifolds, densities).  
- Examples: K-Means, Hierarchical Clustering, DBSCAN, PCA, t-SNE, Autoencoders.  
- Use-cases: customer segmentation, anomaly detection, data exploration, feature learning.

---

## Q2 — How does K-Means clustering algorithm work?

**Answer:**  
K-Means partitions `n` samples into `K` clusters by minimizing within-cluster variance (sum of squared distances to cluster centroids).

**Steps:**
1. **Initialize** `K` centroids (randomly or with K-Means++).  
2. **Assign** each point to its nearest centroid.  
3. **Update** centroids as the mean of assigned points.  
4. **Repeat** until convergence (centroids no longer move significantly).

**Properties:**
- Objective: minimize inertia (sum of squared distances to centroids).  
- Works best for spherical, equally sized clusters.  
- Sensitive to initialization and scaling.

---

## Q3 — Explain the concept of a dendrogram in hierarchical clustering.

**Answer:**  
A **dendrogram** is a tree-like diagram that visualizes how clusters are merged (agglomerative) or split (divisive).

**Reading a dendrogram:**
- Leaves represent individual data points.  
- Height of merge → distance or dissimilarity.  
- Cutting horizontally at a chosen height determines the number of clusters.  
- Taller merges mean less similarity between clusters.

---

## Q4 — What is the main difference between K-Means and Hierarchical Clustering?

**Answer:**

**K-Means:**
- Requires pre-defined `K`.  
- Produces one partition (flat clustering).  
- Fast and scalable.

**Hierarchical Clustering:**
- Builds a hierarchy of clusters (dendrogram).  
- Does not require specifying `K` beforehand.  
- Captures nested structures, but slower and more memory-intensive.

---

## Q5 — What are the advantages of DBSCAN over K-Means?

**Answer:**
- Does **not require specifying K** beforehand.  
- Detects clusters of **arbitrary shapes**.  
- Robust to **noise and outliers**.  
- Handles clusters of varying sizes and densities.

**Limitations:** struggles with clusters having very different densities; sensitive to `eps` and `min_samples`.

---

## Q6 — When would you use Silhouette Score in clustering?

**Answer:**  
Use **Silhouette Score** to measure how well data points are clustered (balance between cohesion and separation).

**Applications:**
- Compare different algorithms or cluster counts.  
- Choose optimal `K` (higher silhouette → better separation).  
- Detect poorly clustered samples (negative values).

---

## Q7 — What are the limitations of Hierarchical Clustering?

**Answer:**
- **Scalability:** O(n²) or worse for large datasets.  
- **Greedy nature:** early wrong merges cannot be undone.  
- **Sensitive to distance/linkage metrics.**  
- **Poor handling of noise/outliers.**  
- Limited support for mixed data types without special measures.

---

## Q8 — Why is feature scaling important in clustering algorithms like K-Means?

**Answer:**  
Because distance-based algorithms are affected by magnitude differences across features.

**Reasons:**
- Features with large ranges dominate distances.  
- Scaling ensures equal contribution of all features.  
- Common scalers: `StandardScaler`, `MinMaxScaler`.

---

## Q9 — How does DBSCAN identify noise points?

**Answer:**  
DBSCAN classifies points based on density using two parameters:  
- **`eps`** — neighborhood radius.  
- **`min_samples`** — minimum points to form a dense region.

**Classification:**
- **Core point:** ≥ `min_samples` within `eps`.  
- **Border point:** within `eps` of a core point but not core itself.  
- **Noise point:** neither core nor border (isolated).

---

## Q10 — Define inertia in the context of K-Means.

**Answer:**  
Inertia (Within-Cluster Sum of Squares, WCSS) measures compactness:

\[
\text{Inertia} = \sum_{k=1}^{K}\sum_{x \in C_k}\|x - \mu_k\|^2
\]

Lower inertia → tighter clusters.  
However, inertia always decreases as `K` increases, so it cannot alone determine optimal `K`.

---

## Q11 — What is the elbow method in K-Means clustering?

**Answer:**  
A heuristic to choose `K`.

**Steps:**
1. Compute inertia for various `K` values.  
2. Plot inertia vs `K`.  
3. The "elbow" point (where inertia drop slows) suggests optimal `K`.

Combining with Silhouette Score gives more reliable results.

---

## Q12 — Describe the concept of “density” in DBSCAN.

**Answer:**  
Density = number of points within a specified radius (`eps`).  
A region is **dense** if it contains at least `min_samples` points.

DBSCAN forms clusters as continuous dense regions separated by sparse areas.

---

## Q13 — Can hierarchical clustering be used on categorical data?

**Answer:**  
Yes, but requires suitable distance metrics:  
- **Hamming distance**, **Jaccard distance**, or **Gower distance**.  
- One-hot encoding can convert categorical data to numerical.  
- Linkage criteria still apply once distances are computed.

---

## Q14 — What does a negative Silhouette Score indicate?

**Answer:**  
A negative Silhouette Score means:
- A point is closer to another cluster than its own → **misclustered**.  
- Average negative score indicates poor separation or overlapping clusters.

**Ranges:**  
- +1 → well-clustered.  
- 0 → borderline.  
- -1 → incorrect cluster assignment.

---

## Q15 — Explain the term “linkage criteria” in hierarchical clustering.

**Answer:**  
Linkage determines **how distances between clusters** are measured when merging.

**Types:**
- **Single linkage:** min distance between members.  
- **Complete linkage:** max distance between members.  
- **Average linkage:** mean pairwise distance.  
- **Ward’s linkage:** minimizes total within-cluster variance.

Choice of linkage affects cluster shape and structure.

---

## Q16 — Why might K-Means clustering perform poorly on data with varying cluster sizes or densities?

**Answer:**  
Because it assumes:
- Spherical, equally sized, and equally dense clusters.

**Problems:**
- Large clusters may be split.  
- Uneven densities distort centroids.  
- Outliers can shift cluster centers.

Use **DBSCAN** or **Gaussian Mixture Models** for uneven clusters.

---

## Q17 — What are the core parameters in DBSCAN, and how do they influence clustering?

**Answer:**  

1. **`eps` (ε):**  
   - Radius defining a neighborhood.  
   - Small `eps` → more clusters, more noise.  
   - Large `eps` → fewer clusters, possible over-merging.

2. **`min_samples`:**  
   - Minimum number of points to form a dense region.  
   - Small → more clusters, less noise.  
   - Large → stricter density, more points labeled as noise.

**Tip:** Use k-distance plots to estimate `eps`.

---

## Q18 — How does K-Means++ improve upon standard K-Means initialization?

**Answer:**  
K-Means++ selects better initial centroids:

**Steps:**
1. Pick first centroid randomly.  
2. Next centroids chosen with probability proportional to squared distance (`D²`) from existing centroids.  
3. Repeat until `K` centroids chosen, then run normal K-Means.

**Benefits:** faster convergence and better cluster quality.

---

## Q19 — What is agglomerative clustering?

**Answer:**  
Agglomerative clustering = bottom-up approach.

**Process:**
- Start with each data point as its own cluster.  
- Iteratively merge the two nearest clusters based on linkage criteria.  
- Continue until all points merge or desired clusters achieved.  

Produces a **dendrogram** showing merge hierarchy.

---

## Q20 — What makes Silhouette Score a better metric than just inertia for model evaluation?

**Answer:**  
- **Inertia** measures only compactness (within-cluster).  
- **Silhouette** measures both **cohesion** (within) and **separation** (between clusters).  

**Advantages of Silhouette:**
- Normalized between -1 and +1.  
- Penalizes poor separation.  
- Detects misclassified samples.  
- Helps select optimal `K` more reliably than inertia.

---

```python
# 29. Create synthetic data using make_blobs and evaluate silhouette scores for k = 2 to 5. Display as a bar chart.
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
import numpy as np

X, y = make_blobs(n_samples=500, centers=4, cluster_std=0.6, random_state=42)
ks = [2,3,4,5]
scores = []
for k in ks:
    km = KMeans(n_clusters=k, random_state=42).fit(X)
    lab = km.labels_
    scores.append(silhouette_score(X, lab))

plt.figure(figsize=(6,4))
plt.bar([str(k) for k in ks], scores)
plt.xlabel('k')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Scores for k=2..5')
plt.show()
```

```python
# 30. Load the Iris dataset and use hierarchical clustering to group data. Plot a dendrogram with average linkage.
from sklearn.datasets import load_iris
from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib.pyplot as plt

iris = load_iris()
X = iris.data
Z = linkage(X, method='average')

plt.figure(figsize=(10, 5))
dendrogram(Z, labels=iris.target, leaf_rotation=90)
plt.title("Hierarchical Clustering Dendrogram (average linkage) - Iris")
plt.xlabel("Sample index")
plt.ylabel("Distance")
plt.show()
```

```python
# 31. Generate synthetic data with overlapping clusters using make_blobs, then apply K-Means and visualize decision boundaries.
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt

X, y = make_blobs(n_samples=400, centers=[[0,0],[2,2],[4,0]], cluster_std=1.4, random_state=42)
kmeans = KMeans(n_clusters=3, random_state=42).fit(X)

# decision boundary grid
xx, yy = np.meshgrid(np.linspace(X[:,0].min()-1,X[:,0].max()+1,200),
                     np.linspace(X[:,1].min()-1,X[:,1].max()+1,200))
grid = np.c_[xx.ravel(), yy.ravel()]
pred = kmeans.predict(grid).reshape(xx.shape)

plt.figure(figsize=(7,5))
plt.contourf(xx, yy, pred, alpha=0.2, cmap='tab10')
plt.scatter(X[:,0], X[:,1], c=kmeans.labels_, cmap='tab10', edgecolor='k', s=30)
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], c='black', marker='X', s=150)
plt.title("K-Means with decision boundaries (overlapping blobs)")
plt.show()
```

```python
# 46. Load the Wine dataset, apply KMeans, and visualize the cluster assignments in a Seaborn pairplot (first 4 features).
from sklearn.datasets import load_wine
from sklearn.cluster import KMeans
import seaborn as sns
import pandas as pd

data = load_wine()
df = pd.DataFrame(data.data, columns=data.feature_names)
df_sub = df.iloc[:, :4]  # first 4 features

kmeans = KMeans(n_clusters=3, random_state=42).fit(df_sub)
df_sub['cluster'] = kmeans.labels_.astype(str)

sns.pairplot(df_sub, hue='cluster', corner=True, diag_kind='kde')
plt.suptitle("Wine dataset - KMeans clusters (first 4 features)", y=1.02)
plt.show()
```

```python
# 47. Generate noisy blobs using make_blobs and use DBSCAN to identify both clusters and noise points. Print the count.
from sklearn.datasets import make_blobs
from sklearn.cluster import DBSCAN
import numpy as np

X, y = make_blobs(n_samples=500, centers=4, cluster_std=1.5, random_state=42)
# add some random noise points
rng = np.random.RandomState(42)
noise = rng.uniform(low=-10, high=10, size=(30,2))
X_noisy = np.vstack([X, noise])

db = DBSCAN(eps=0.8, min_samples=8)
labels = db.fit_predict(X_noisy)
(unique, counts) = np.unique(labels, return_counts=True)
print("Label counts (including -1 = noise):", dict(zip(unique, counts)))
print("Number of noise points:", int(np.sum(labels == -1)))
```

```python
# 48. Load the Digits dataset, reduce dimensions using t-SNE, then apply Agglomerative Clustering and plot the clusters.
from sklearn.datasets import load_digits
from sklearn.manifold import TSNE
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt

digits = load_digits()
X = digits.data
tsne = TSNE(n_components=2, random_state=42, init='pca', learning_rate='auto', perplexity=30)
X2 = tsne.fit_transform(X)

agg = AgglomerativeClustering(n_clusters=10)
labels = agg.fit_predict(X2)

plt.figure(figsize=(8,6))
plt.scatter(X2[:,0], X2[:,1], c=labels, cmap='tab10', s=25, edgecolor='k')
plt.title("Agglomerative Clustering on Digits (t-SNE -> 2D)")
plt.show()
```
