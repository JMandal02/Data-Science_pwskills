**1. What is NumPy, and why is it widely used in Python?**

- NumPy (Numerical Python) is a Python library for efficient numerical computing.
Provides n-dimensional arrays (ndarray) which are faster and more memory-efficient than Python lists.
Supports vectorized operations, linear algebra, random number generation, and broadcasting.
Widely used because it is the foundation of the scientific Python ecosystem (Pandas, Scikit-learn, TensorFlow all rely on it).

**2. How does broadcasting work in NumPy?**

- Broadcasting is NumPy’s ability to perform operations on arrays of different shapes by automatically expanding smaller arrays to match larger ones without creating unnecessary copies, enabling efficient computations.

**3. What is a Pandas DataFrame?**

- A Pandas DataFrame is a two-dimensional, labeled data structure similar to a table in SQL or Excel. It stores data in rows and columns, supports different data types, and provides powerful tools for data manipulation and analysis.

**4. Explain the use of the groupby() method in Pandas.**

- The groupby() method in Pandas is used to split data into groups based on some criteria, apply a function (like mean, sum, count), and combine the results, enabling efficient aggregation and analysis.

**5. Why is Seaborn preferred for statistical visualizations?**

- Seaborn is preferred because it provides high-level, easy-to-use functions for creating attractive and informative statistical graphics. It has built-in support for data frames, beautiful default styles, and specialized statistical plots like heatmaps, pairplots, and violin plots.

**6. What are the differences between NumPy arrays and Python lists?**

- NumPy arrays are faster, memory-efficient, and support vectorized operations, while Python lists are slower, consume more memory, and require loops for element-wise operations. NumPy arrays are homogeneous, whereas lists can store mixed data types.

**7. What is a heatmap, and when should it be used?**

- A heatmap is a graphical representation of data where values are represented by colors. It is mainly used to visualize correlations, frequency distributions, and patterns in large datasets where numerical values are difficult to interpret directly.

**8. What does the term “vectorized operation” mean in NumPy?**

- Vectorized operation means applying a mathematical operation to an entire array at once without explicit loops. This leads to faster execution and cleaner code compared to iterative approaches.

**9. How does Matplotlib differ from Plotly?**

- Matplotlib creates static, highly customizable plots suitable for publications, while Plotly provides interactive, web-based visualizations that allow zooming, hovering, and dynamic exploration, making it ideal for dashboards.

**10. What is the significance of hierarchical indexing in Pandas?**

- Hierarchical indexing (MultiIndex) allows multiple levels of indexing in rows or columns. It is significant because it enables the representation of higher-dimensional data in a 2D structure, making complex datasets easier to handle.

**11. What is the role of Seaborn's pairplot() function?**

- pairplot() generates a grid of scatterplots showing relationships between all pairs of numerical variables in a dataset. It is useful for exploring correlations and distributions quickly.

**12. What is the purpose of the describe() function in Pandas?**

- The describe() function provides summary statistics of numerical columns in a dataset, including count, mean, standard deviation, minimum, quartiles, and maximum, helping in quick exploratory analysis.

**13. Why is handling missing data important in Pandas?**

- Handling missing data is important because unaddressed gaps can lead to incorrect calculations, biased results, or errors during analysis. Pandas offers methods like dropna() and fillna() to manage missing values effectively.

**14. What are the benefits of using Plotly for data visualization?**

- Plotly offers interactive, web-based visualizations with features like hover information, zooming, and dynamic updates. It is highly suitable for dashboards, supports multiple chart types, and integrates well with Python notebooks and web apps.

**15. How does NumPy handle multidimensional arrays?**

- NumPy handles multidimensional arrays using the ndarray object, which can store data in any number of dimensions. It allows operations across different axes, enabling efficient representation and manipulation of complex datasets.

**16. What is the role of Bokeh in data visualization?**

- Bokeh is a Python library for creating interactive, browser-based visualizations. It allows real-time data streaming, supports large datasets, and is commonly used for building dashboards and web applications.

**17. Explain the difference between apply() and map() in Pandas.**

- map() works only on Pandas Series and applies a function element-wise.
apply() works on both Series and DataFrames, applying functions across rows or columns, making it more flexible.

**18. What are some advanced features of NumPy?**

- Advanced features of NumPy include broadcasting, vectorized operations, linear algebra functions, Fourier transforms, random number generation, memory mapping for large datasets, and masked arrays for handling missing values.

**19. How does Pandas simplify time series analysis?**

- Pandas simplifies time series analysis through features like datetime indexing, resampling (daily, monthly, yearly), rolling windows, shifting, and powerful time-based slicing, making it easy to manipulate and analyze temporal data.

**20. What is the role of a pivot table in Pandas?**

- A pivot table summarizes and reorganizes data by applying aggregation functions (sum, mean, count) across two or more categorical dimensions. It is useful for analyzing grouped data efficiently.

**21. Why is NumPy's array slicing faster than Python's list slicing?**

- NumPy slicing is faster because arrays are stored in contiguous memory blocks, and slices are views (not copies) of the original data, while Python lists require copying elements one by one.

**22. What are some common use cases for Seaborn?**

Correlation heatmaps
Distribution plots (histogram, KDE, violin)
Categorical plots (bar, box, swarm)
Regression analysis plots (regplot, lmplot)
Pairwise relationship visualization (pairplot)

```python
#2: Write a Pandas script to find the mean of a specific column in a DataFrame
import pandas as pd

data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [25, 30, 35, 40],
    'Salary': [50000, 60000, 75000, 80000]
}

df = pd.DataFrame(data)

mean_salary = df['Salary'].mean()

print("Mean of Salary column:", mean_salary)
```

```python
#4: How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = {
    'Math': [90, 80, 85, 70, 95],
    'Science': [88, 78, 84, 72, 92],
    'English': [75, 85, 80, 70, 90],
    'History': [70, 72, 68, 65, 75]
}

df = pd.DataFrame(data)

corr_matrix = df.corr()

print("Correlation Matrix:\n", corr_matrix)

plt.figure(figsize=(6,4))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", linewidths=0.5)

plt.title("Correlation Matrix Heatmap")
plt.show()
```

```python
#5: Generate a bar plot using Plotly
import plotly.express as px
import pandas as pd

data = {
    "Fruits": ["Apple", "Banana", "Orange", "Grapes", "Mango"],
    "Sales": [50, 30, 40, 20, 60]
}

df = pd.DataFrame(data)

fig = px.bar(df, x="Fruits", y="Sales", title="Fruit Sales", color="Fruits")

fig.show()
```

```python
#6: Create a DataFrame and add a new column based on an existing column
import pandas as pd

data = {
    "Name": ["Alice", "Bob", "Charlie", "David"],
    "Age": [25, 30, 35, 40]
}

df = pd.DataFrame(data)

print("Original DataFrame:\n", df)

df["Age_Group"] = df["Age"].apply(lambda x: "Young" if x < 30 else "Adult")

print("\nDataFrame with New Column:\n", df)
```

```python
#8: Create a line plot with multiple lines using Matplotlib
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]
y2 = [1, 3, 5, 7, 9]
y3 = [2, 3, 4, 5, 6]

plt.plot(x, y1, label="Line 1", color="blue", marker="o")
plt.plot(x, y2, label="Line 2", color="red", linestyle="--", marker="s")
plt.plot(x, y3, label="Line 3", color="green", linestyle="-.", marker="^")

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Line Plot with Multiple Lines")

plt.legend()

plt.show()
```

```python
#9: Generate a Pandas DataFrame and filter rows where a column value is greater than a threshold
import pandas as pd

data = {
    "Name": ["Alice", "Bob", "Charlie", "David", "Eve"],
    "Age": [25, 30, 35, 40, 28],
    "Salary": [50000, 60000, 75000, 80000, 55000]
}

df = pd.DataFrame(data)

print("Original DataFrame:\n", df)

filtered_df = df[df["Salary"] > 60000]

print("\nFiltered DataFrame (Salary > 60000):\n", filtered_df)
```

```python
#11: Perform matrix multiplication using NumPy
import numpy as np

A = np.array([[1, 2, 3],
              [4, 5, 6]])

B = np.array([[7, 8],
              [9, 10],
              [11, 12]])

result = np.dot(A, B)

print("Matrix Multiplication Result:\n", result)
```

```python
#13: Create a 3D scatter plot using Plotly.
import plotly.express as px
import pandas as pd

data = {
    "X": [1, 2, 3, 4, 5],
    "Y": [10, 11, 12, 13, 14],
    "Z": [5, 6, 7, 8, 9],
    "Category": ["A", "B", "A", "B", "A"]
}

df = pd.DataFrame(data)

fig = px.scatter_3d(df, x="X", y="Y", z="Z",
                    color="Category",
                    size="Z",
                    symbol="Category",
                    title="3D Scatter Plot with Plotly")

fig.show()
```
