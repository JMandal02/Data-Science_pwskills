## Question 1
**Compare and contrast NLTK and spaCy in terms of features, ease of use, and performance.**

### Answer

**NLTK (Natural Language Toolkit):**
- Built for education and research purposes
- Very large collection of corpora, lexicons, and tools
- Supports many algorithms but requires more manual setup
- Slower in processing large text data
- More flexible — good for experimenting and learning
- Example tasks: tokenization, stemming, chunking, parsing

**spaCy:**
- Built for production and real-world applications
- Faster and more efficient than NLTK
- Uses pre-trained models out of the box
- Cleaner and simpler API — less code needed
- Supports deep learning integration natively
- Example tasks: NER, dependency parsing, lemmatization

**Key Differences:**

| Feature | NLTK | spaCy |
|---------|------|-------|
| Purpose | Education / Research | Production / Industry |
| Speed | Slow | Fast |
| Ease of Use | Moderate | Easy |
| Pre-trained Models | Limited | Yes (many languages) |
| Deep Learning Support | No | Yes |
| Best For | Learning NLP concepts | Building NLP applications |

**When to use which:**
- Use NLTK → when learning NLP or doing academic research
- Use spaCy → when building real applications that need speed and accuracy

## Question 2
**What is TextBlob and how does it simplify common NLP tasks like sentiment analysis and translation?**

### Answer

**What is TextBlob:**
- Simple Python library built on top of NLTK and Pattern
- Designed for beginners — very easy to use
- Provides a clean API for common NLP tasks
- Works with just a few lines of code

**What TextBlob can do:**
- Tokenization → split text into words and sentences
- POS Tagging → assign part-of-speech to each word
- Noun Phrase Extraction → find noun phrases in text
- Sentiment Analysis → returns polarity and subjectivity scores
- Spelling Correction → auto-corrects misspelled words
- Translation → translate text to another language (uses Google Translate)
- Lemmatization → reduce words to base form
- N-grams → generate word n-grams

**How it simplifies Sentiment Analysis:**
- Returns two scores automatically:
  - Polarity → ranges from -1.0 (negative) to +1.0 (positive)
  - Subjectivity → ranges from 0.0 (objective) to 1.0 (subjective)
- Just one line: `TextBlob(text).sentiment`

**How it simplifies Translation:**
- Just one line: `TextBlob(text).translate(to='fr')` for French
- No need to call external APIs manually

**Limitation:**
- Not suitable for production — slower and less accurate than spaCy or transformers
- Good only for quick prototyping and learning

## Question 4
**Describe the architecture and functioning of a Recurrent Neural Network (RNN).**

### Answer

**What is an RNN:**
- A type of neural network designed for sequential data
- Unlike normal networks, it has memory — it remembers previous inputs
- Used for tasks where order of data matters like text and speech

**Architecture:**
- Has three main components:
  - Input Layer → receives one token/word at a time
  - Hidden Layer → has a loop that passes output back as input
  - Output Layer → produces prediction at each or final time step
- At each time step t:
  - Takes current input x(t)
  - Takes previous hidden state h(t-1)
  - Produces new hidden state h(t)
  - Formula: h(t) = tanh(W * x(t) + U * h(t-1) + b)

**How it works step by step:**
- Step 1 → Feed word 1, get hidden state h1
- Step 2 → Feed word 2 + h1, get hidden state h2
- Step 3 → Feed word 3 + h2, get hidden state h3
- Final output is based on last hidden state

**Use cases in NLP:**
- Text classification
- Language modeling
- Machine translation
- Speech recognition
- Named entity recognition

**Problem with basic RNN:**
- Vanishing Gradient Problem → early words are forgotten in long sequences
- Cannot handle long-term dependencies well
- This is exactly why LSTM and GRU were created to fix this

## Question 5
**What is the key difference between LSTM and GRU networks in NLP applications?**

### Answer

**LSTM (Long Short-Term Memory):**
- Designed to fix the vanishing gradient problem of RNNs
- Has 3 gates to control information flow:
  - Forget Gate → decides what to forget from past
  - Input Gate → decides what new info to store
  - Output Gate → decides what to output
- Has two states:
  - Cell State → carries long-term memory
  - Hidden State → carries short-term memory
- More parameters → slower to train
- Better for complex tasks with very long sequences

**GRU (Gated Recurrent Unit):**
- A simplified and faster version of LSTM
- Has only 2 gates:
  - Reset Gate → how much past info to forget
  - Update Gate → how much past info to carry forward
- Only one hidden state (no separate cell state)
- Fewer parameters → faster to train
- Performs similarly to LSTM on most tasks

**Key Differences:**

| Feature | LSTM | GRU |
|---------|------|-----|
| Gates | 3 (forget, input, output) | 2 (reset, update) |
| States | Cell state + Hidden state | Only hidden state |
| Parameters | More | Less |
| Speed | Slower | Faster |
| Memory | Better for long sequences | Good for shorter sequences |
| Complexity | Higher | Lower |

**When to use which:**
- Use LSTM → when sequence is very long and complex (e.g., document classification)
- Use GRU → when faster training needed with good results (e.g., chatbots, short text)
- In practice GRU often gives similar accuracy to LSTM but trains faster

```python
# Question 6
# Write a Python program using TextBlob to perform sentiment analysis on the
# following paragraph of text:
#
# "I had a great experience using the new mobile banking app. The interface is intuitive,
# and customer support was quick to resolve my issue. However, the app did crash once
# during a transaction, which was frustrating"
#
# Your program should print out the polarity and subjectivity scores.

from textblob import TextBlob

text = (
    "I had a great experience using the new mobile banking app. "
    "The interface is intuitive, and customer support was quick to resolve my issue. "
    "However, the app did crash once during a transaction, which was frustrating"
)

blob = TextBlob(text)

polarity     = blob.sentiment.polarity
subjectivity = blob.sentiment.subjectivity

print("Overall Sentiment:")
print(f"  Polarity Score    : {polarity:.4f}  (-1.0 = negative, +1.0 = positive)")
print(f"  Subjectivity Score: {subjectivity:.4f}  (0.0 = objective, 1.0 = subjective)")

if polarity > 0:
    label = "Positive"
elif polarity < 0:
    label = "Negative"
else:
    label = "Neutral"

print(f"  Sentiment Label   : {label}")

# Sentence-wise breakdown
print("\nSentence-wise Sentiment:")
print(f"  {'Sentence':<65} {'Polarity':>10} {'Subjectivity':>14}")
print("-" * 95)
for sentence in blob.sentences:
    print(f"  {str(sentence):<65} {sentence.sentiment.polarity:>10.4f} {sentence.sentiment.subjectivity:>14.4f}")
```

```python
# Question 7
# Given the sample paragraph below, perform string tokenization and
# frequency distribution using Python and NLTK:
#
# "Natural Language Processing (NLP) is a fascinating field that combines linguistics,
# computer science, and artificial intelligence. It enables machines to understand,
# interpret, and generate human language. Applications of NLP include chatbots,
# sentiment analysis, and machine translation. As technology advances, the role of NLP
# in modern solutions is becoming increasingly critical."

import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.probability import FreqDist
from nltk.corpus import stopwords
import string
import matplotlib.pyplot as plt

nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')

paragraph = (
    "Natural Language Processing (NLP) is a fascinating field that combines linguistics, "
    "computer science, and artificial intelligence. It enables machines to understand, "
    "interpret, and generate human language. Applications of NLP include chatbots, "
    "sentiment analysis, and machine translation. As technology advances, the role of NLP "
    "in modern solutions is becoming increasingly critical."
)

# 1. Sentence Tokenization
print("1. Sentence Tokenization:")
for i, s in enumerate(sent_tokenize(paragraph), 1):
    print(f"   {i}. {s}")

# 2. Word Tokenization
words = word_tokenize(paragraph)
print(f"\n2. Word Tokenization:")
print(f"   Total tokens: {len(words)}")
print(f"   Tokens: {words}")

# 3. Remove stop words and punctuation
stop_words = set(stopwords.words('english'))
clean = [w.lower() for w in words if w.lower() not in stop_words and w not in string.punctuation]
print(f"\n3. After Cleaning:")
print(f"   Total: {len(clean)}")
print(f"   {clean}")

# 4. Frequency Distribution
fdist = FreqDist(clean)
print("\n4. Top 10 Frequent Words:")
print(f"   {'Word':<20} {'Count'}")
print(f"   {'-'*28}")
for word, freq in fdist.most_common(10):
    print(f"   {word:<20} {freq}")

# 5. Plot
fdist.plot(10, title="Top 10 Words - Frequency Distribution")
plt.tight_layout()
plt.show()
```

```python
# Question 8
# Implement a basic LSTM model in Keras for a text classification task using
# the following dummy dataset. Your model should classify sentences as either
# positive (1) or negative (0).
#
# Dataset:
# texts = [
#     "I love this project",           # Positive
#     "This is an amazing experience", # Positive
#     "I hate waiting in line",        # Negative
#     "This is the worst service",     # Negative
#     "Absolutely fantastic!"          # Positive
# ]
# labels = [1, 1, 0, 0, 1]
#
# Preprocess the text, tokenize it, pad sequences, and build an LSTM model
# to train on this data. Use Keras with TensorFlow backend.

import numpy as np
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense

# Dataset
texts = [
    "I love this project",
    "This is an amazing experience",
    "I hate waiting in line",
    "This is the worst service",
    "Absolutely fantastic!"
]
labels = [1, 1, 0, 0, 1]

# Step 1 — Tokenize
tokenizer = Tokenizer()
tokenizer.fit_on_texts(texts)
sequences = tokenizer.texts_to_sequences(texts)
vocab_size = len(tokenizer.word_index) + 1

print("Step 1 — Tokenization:")
print(f"  Vocab size: {vocab_size}")
print(f"  Sequences: {sequences}")

# Step 2 — Pad sequences
max_len = max(len(s) for s in sequences)
X = pad_sequences(sequences, maxlen=max_len, padding='post')
y = np.array(labels)

print(f"\nStep 2 — Padded Sequences (max_len={max_len}):")
print(X)

# Step 3 — Build LSTM Model
model = Sequential([
    Embedding(input_dim=vocab_size, output_dim=8, input_length=max_len),
    LSTM(16),
    Dense(1, activation='sigmoid')
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.summary()

# Step 4 — Train
print("\nStep 4 — Training:")
model.fit(X, y, epochs=20, verbose=1)

# Step 5 — Predict
print("\nStep 5 — Predictions:")
predictions = model.predict(X)
for text, pred, actual in zip(texts, predictions, labels):
    label = "Positive" if pred[0] >= 0.5 else "Negative"
    print(f"  '{text}'")
    print(f"    Predicted: {label} ({pred[0]:.4f}) | Actual: {'Positive' if actual==1 else 'Negative'}")
```

```python
# Question 9
# Using spaCy, build a simple NLP pipeline that includes tokenization,
# lemmatization, and entity recognition. Use the following paragraph as dataset:
#
# "Homi Jehangir Bhaba was an Indian nuclear physicist who played a key role in the
# development of India's atomic energy program. He was the founding director of the Tata
# Institute of Fundamental Research (TIFR) and was instrumental in establishing the
# Atomic Energy Commission of India."
#
# Write a Python program that processes this text using spaCy, then prints
# tokens, their lemmas, and any named entities found.

# !pip install spacy
# !python -m spacy download en_core_web_sm

import spacy

nlp = spacy.load("en_core_web_sm")

text = (
    "Homi Jehangir Bhaba was an Indian nuclear physicist who played a key role in the "
    "development of India's atomic energy program. He was the founding director of the Tata "
    "Institute of Fundamental Research (TIFR) and was instrumental in establishing the "
    "Atomic Energy Commission of India."
)

doc = nlp(text)

# 1. Tokenization + Lemmatization
print("1. Tokens and Lemmas:")
print(f"   {'Token':<25} {'Lemma':<25} {'POS'}")
print(f"   {'-'*65}")
for token in doc:
    if not token.is_space:
        print(f"   {token.text:<25} {token.lemma_:<25} {token.pos_}")

# 2. Named Entity Recognition
print("\n2. Named Entities Found:")
print(f"   {'Entity':<35} {'Label':<15} {'Description'}")
print(f"   {'-'*70}")
for ent in doc.ents:
    print(f"   {ent.text:<35} {ent.label_:<15} {spacy.explain(ent.label_)}")

# 3. Sentences
print("\n3. Sentences:")
for i, sent in enumerate(doc.sents, 1):
    print(f"   {i}. {sent.text.strip()}")
```

```python
# Question 10
# You are working on a chatbot for a mental health platform. Explain how you
# would leverage LSTM or GRU networks along with libraries like spaCy or
# Stanford NLP to understand and respond to user input effectively.
# Detail your architecture, data preprocessing pipeline, and any ethical
# considerations.
#
# Architecture:
# - Input Layer        → user types a message
# - Preprocessing      → spaCy cleans, lemmatizes, removes stop words
# - Embedding Layer    → converts words to dense vectors
# - LSTM/GRU Layer     → understands sequence and context of words
# - Dense Output Layer → classifies user intent (distressed / positive)
# - Response Module    → maps intent to appropriate response
#
# Preprocessing Pipeline:
# - Step 1 → collect labeled mental health conversation dataset
# - Step 2 → clean text using spaCy (lemmatize, remove noise)
# - Step 3 → tokenize and convert to integer sequences
# - Step 4 → pad sequences to fixed length
# - Step 5 → train LSTM/GRU on intent classification
# - Step 6 → map predicted intent to response templates
#
# Ethical Considerations:
# - Never store user conversations without consent
# - Always clarify the bot is not a replacement for a real therapist
# - Add crisis detection — escalate to human if suicidal signals detected
# - Avoid biased training data
# - Follow data privacy laws (GDPR, HIPAA)

import numpy as np
import spacy
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense

nlp = spacy.load("en_core_web_sm")

# Sample dataset
texts = [
    "I feel very anxious today",
    "I am so stressed about everything",
    "I can't sleep and feel hopeless",
    "I feel happy and calm",
    "Things are going well for me",
    "I am feeling overwhelmed",
    "I had a great day today",
    "I feel like no one understands me"
]
labels = [1, 1, 1, 0, 0, 1, 0, 1]   # 1 = distressed, 0 = positive

# Step 1 — Preprocess using spaCy
def preprocess(text):
    doc = nlp(text.lower())
    return " ".join([token.lemma_ for token in doc if not token.is_stop and not token.is_punct])

clean_texts = [preprocess(t) for t in texts]
print("Step 1 — Preprocessed Texts:")
for orig, clean in zip(texts, clean_texts):
    print(f"  Original : {orig}")
    print(f"  Cleaned  : {clean}\n")

# Step 2 — Tokenize and pad
tokenizer = Tokenizer()
tokenizer.fit_on_texts(clean_texts)
sequences = tokenizer.texts_to_sequences(clean_texts)
max_len = max(len(s) for s in sequences)
X = pad_sequences(sequences, maxlen=max_len, padding='post')
y = np.array(labels)

print(f"Step 2 — Vocab size: {len(tokenizer.word_index)+1} | Max length: {max_len}")

# Step 3 — Build LSTM model
model = Sequential([
    Embedding(input_dim=len(tokenizer.word_index)+1, output_dim=16, input_length=max_len),
    LSTM(32),
    Dense(1, activation='sigmoid')
])
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.summary()

# Step 4 — Train
print("\nStep 4 — Training:")
model.fit(X, y, epochs=30, verbose=0)
print("Training complete!")

# Step 5 — Predict and respond
def predict_intent(user_input):
    cleaned = preprocess(user_input)
    seq = tokenizer.texts_to_sequences([cleaned])
    padded = pad_sequences(seq, maxlen=max_len, padding='post')
    score = model.predict(padded, verbose=0)[0][0]
    intent = "distressed" if score >= 0.5 else "positive"
    return intent, score

responses = {
    "distressed": "I hear you. It sounds like things are tough right now. Would you like to talk more?",
    "positive"  : "That's great to hear! Keep taking care of yourself."
}

print("\nStep 5 — Chatbot Test:")
test_inputs = [
    "I am feeling really down lately",
    "I am doing great, feeling good",
    "Everything feels too much to handle"
]

for user_input in test_inputs:
    intent, score = predict_intent(user_input)
    print(f"\n  User  : {user_input}")
    print(f"  Intent: {intent} (score: {score:.4f})")
    print(f"  Bot   : {responses[intent]}")

print("\n--- Ethical Note ---")
print("This is a prototype only.")
print("Always refer users in crisis to a licensed professional or helpline.")
```
