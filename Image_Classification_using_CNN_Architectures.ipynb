## **Question 1: What is a Convolutional Neural Network (CNN), and how does it differ from traditional fully connected neural networks in terms of architecture and performance on image data?**

### **Answer:**

A **Convolutional Neural Network (CNN)** is a specialized class of deep neural networks designed primarily for processing **grid-like data**, such as images. CNNs automatically learn spatial hierarchies of features through convolution operations, making them highly effective for image-related tasks such as classification, detection, and segmentation.

### **Architecture of CNN**

A typical CNN consists of the following layers:

1. **Convolutional Layer** – Applies learnable filters to extract local features like edges, corners, and textures.  
2. **Activation Function (ReLU)** – Introduces non-linearity to help the model learn complex patterns.  
3. **Pooling Layer** – Reduces spatial dimensions, lowering computational cost and overfitting.  
4. **Fully Connected Layer** – Combines extracted features to perform classification.  
5. **Softmax Layer** – Outputs class probabilities.

### **Difference Between CNN and Fully Connected Neural Networks**

| Aspect | CNN | Fully Connected Neural Network |
|------|-----|--------------------------------|
| Connectivity | Local receptive fields | Every neuron connected to all inputs |
| Parameters | Fewer (shared weights) | Very large |
| Spatial Awareness | Preserves spatial structure | Loses spatial information |
| Computation | Efficient | Computationally expensive |
| Performance on Images | High accuracy | Poor scalability |

### **Performance on Image Data**

CNNs outperform traditional fully connected networks because:
- They **preserve spatial relationships**
- They **reduce parameters via weight sharing**
- They **generalize better** on image data

Hence, CNNs are the foundation of modern computer vision systems.


## **Question 2: Discuss the architecture of LeNet-5 and explain how it laid the foundation for modern deep learning models in computer vision. Include references to its original research paper.**

### **Answer:**

**LeNet-5** is one of the earliest and most influential convolutional neural network architectures, proposed by **Yann LeCun and his collaborators in 1998** for handwritten digit recognition.

### **Architecture of LeNet-5**

LeNet-5 consists of the following layers:

1. **Input Layer** – Accepts a 32×32 grayscale image.  
2. **C1 (Convolution Layer)** – Applies 6 convolutional filters of size 5×5 to extract low-level features.  
3. **S2 (Subsampling Layer)** – Performs average pooling to reduce spatial dimensions.  
4. **C3 (Convolution Layer)** – Uses 16 feature maps to capture more complex patterns.  
5. **S4 (Subsampling Layer)** – Further reduces feature map size using average pooling.  
6. **C5 (Fully Connected Convolution Layer)** – Acts as a fully connected layer producing high-level features.  
7. **F6 and Output Layer** – Performs final classification using fully connected neurons.

### **How LeNet-5 Laid the Foundation for Modern CNNs**

LeNet-5 introduced several key concepts that are fundamental to modern deep learning models:
- **Local receptive fields** for spatial feature extraction  
- **Weight sharing** to reduce the number of parameters  
- **Pooling layers** to achieve translation invariance  
- **End-to-end learning** using backpropagation  

These principles later influenced advanced architectures such as AlexNet, VGGNet, and ResNet.

### **Original Research Reference**

LeCun, Y., Bottou, L., Bengio, Y., & Haffner, P.  
**“Gradient-Based Learning Applied to Document Recognition”**, IEEE, 1998.


## **Question 3: Compare and contrast AlexNet and VGGNet in terms of design principles, number of parameters, and performance. Highlight key innovations and limitations of each.**

### **Answer:**

**AlexNet** and **VGGNet** are two landmark convolutional neural network architectures that significantly advanced the field of image classification.

### **Design Principles**

- **AlexNet:**  
  AlexNet employs relatively large convolutional filters (such as 11×11 and 5×5) with a smaller number of layers. It focuses on learning high-level features quickly using fewer but wider layers.

- **VGGNet:**  
  VGGNet follows a deep and uniform architecture using very small 3×3 convolutional filters stacked together. This design increases depth while keeping the receptive field manageable.

### **Number of Parameters**

- **AlexNet:** Approximately **60 million parameters**, making it computationally lighter compared to deeper models.
- **VGGNet:** Around **138 million parameters**, resulting in higher memory and computation requirements.

### **Performance Comparison**

VGGNet generally achieves higher classification accuracy than AlexNet due to its increased depth and better feature representation. However, AlexNet remains faster to train and less resource-intensive.

### **Key Innovations**

- **AlexNet Innovations:**
  - Introduction of **ReLU activation** for faster convergence  
  - Use of **GPU-based training**  
  - Application of **Dropout** to reduce overfitting  

- **VGGNet Innovations:**
  - Use of **small convolutional filters (3×3)**  
  - Deeper network architecture  
  - Uniform layer design for simplicity and effectiveness  

### **Limitations**

- **AlexNet:** Limited depth restricts its ability to capture very complex features.  
- **VGGNet:** Extremely high computational and memory requirements make it inefficient for deployment on resource-constrained systems.

Both architectures contributed significantly to the evolution of deep convolutional neural networks.


## **Question 4: What is transfer learning in the context of image classification? Explain how it helps in reducing computational costs and improving model performance with limited data.**

### **Answer:**

**Transfer learning** is a deep learning technique in which a model trained on a large and diverse dataset is reused for a different but related task. In image classification, transfer learning commonly involves using a pre-trained convolutional neural network and adapting it to a new dataset.

### **Concept of Transfer Learning**

Instead of training a CNN from scratch, a model that has already learned useful visual features such as edges, shapes, and textures is utilized. The earlier layers of the network are retained, while the final classification layers are replaced and retrained for the new task.

### **Reduction of Computational Cost**

Transfer learning reduces computational cost because:
- The model does not need to be trained from the beginning.
- Fewer epochs are required for convergence.
- Only a small number of parameters are updated during training.

This makes it suitable for systems with limited computational resources.

### **Improvement in Performance with Limited Data**

Transfer learning improves performance when labeled data is scarce because:
- Pre-trained models capture general features applicable across domains.
- The risk of overfitting is reduced.
- The model generalizes better even with small datasets.



## **Question 5: Describe the role of residual connections in ResNet architecture. How do they address the vanishing gradient problem in deep CNNs?**

### **Answer:**

**Residual connections** are shortcut pathways introduced in the ResNet architecture that allow the input of a layer to be added directly to its output. These connections enable the network to learn residual mappings instead of directly learning complex transformations.

### **Role of Residual Connections**

In a residual block, instead of learning a direct mapping \( H(x) \), the network learns a residual function \( F(x) \) such that:

\[
H(x) = F(x) + x
\]

This allows the original input \( x \) to bypass one or more layers and be added back to the output.

### **Vanishing Gradient Problem**

In very deep convolutional neural networks, gradients tend to become extremely small during backpropagation. This prevents early layers from learning effectively, leading to slow convergence and degraded performance.

### **How Residual Connections Solve the Problem**

Residual connections address the vanishing gradient problem by:
- Providing a direct path for gradient flow
- Ensuring better information propagation across layers
- Preventing degradation of training accuracy in deep networks


## Question 6  
### Implement the LeNet-5 architecture using TensorFlow or PyTorch to classify the MNIST dataset. Report the accuracy and training time.

### Answer

Below is an implementation of the **LeNet-5 architecture using TensorFlow (Keras API)** to classify handwritten digits from the MNIST dataset.

### LeNet-5 Implementation (TensorFlow)

```python
import torch
import torch.nn as nn
import torch.optim as optim
import torchvision
import torchvision.transforms as transforms
import time

# 1. Prepare Dataset (MNIST)
# LeNet-5 originally takes 32x32 input, so we resize MNIST (28x28) to 32x32
transform = transforms.Compose([
    transforms.Resize((32, 32)),
    transforms.ToTensor(),
    transforms.Normalize((0.1307,), (0.3081,))
])

trainset = torchvision.datasets.MNIST(root='./data', train=True, download=True, transform=transform)
trainloader = torch.utils.data.DataLoader(trainset, batch_size=64, shuffle=True)

testset = torchvision.datasets.MNIST(root='./data', train=False, download=True, transform=transform)
testloader = torch.utils.data.DataLoader(testset, batch_size=64, shuffle=False)

# 2. Define LeNet-5 Architecture
class LeNet5(nn.Module):
    def __init__(self):
        super(LeNet5, self).__init__()
        # Layer C1: Convolutional (Input: 1, Output: 6, Kernel: 5x5)
        self.c1 = nn.Conv2d(1, 6, kernel_size=5)
        # Layer S2: Average Pooling (Kernel: 2x2, Stride: 2)
        self.s2 = nn.AvgPool2d(kernel_size=2, stride=2)
        # Layer C3: Convolutional (Input: 6, Output: 16, Kernel: 5x5)
        self.c3 = nn.Conv2d(6, 16, kernel_size=5)
        # Layer S4: Average Pooling (Kernel: 2x2, Stride: 2)
        self.s4 = nn.AvgPool2d(kernel_size=2, stride=2)
        # Layer C5: Convolutional (Input: 16, Output: 120, Kernel: 5x5)
        self.c5 = nn.Conv2d(16, 120, kernel_size=5)
        # Layer F6: Fully Connected
        self.f6 = nn.Linear(120, 84)
        # Output Layer: Fully Connected
        self.output = nn.Linear(84, 10)

        self.tanh = nn.Tanh()

    def forward(self, x):
        x = self.tanh(self.c1(x))
        x = self.s2(x)
        x = self.tanh(self.c3(x))
        x = self.s4(x)
        x = self.tanh(self.c5(x))
        x = torch.flatten(x, 1) # Flatten for dense layers
        x = self.tanh(self.f6(x))
        x = self.output(x)
        return x

# 3. Training Configuration
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = LeNet5().to(device)
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

# 4. Training Loop
print(f"Starting training on {device}...")
start_time = time.time()
epochs = 5

for epoch in range(epochs):
    model.train()
    running_loss = 0.0
    for images, labels in trainloader:
        images, labels = images.to(device), labels.to(device)

        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        running_loss += loss.item()

    print(f"Epoch {epoch+1} - Loss: {running_loss/len(trainloader):.4f}")

end_time = time.time()
training_duration = end_time - start_time

# 5. Evaluation
model.eval()
correct = 0
total = 0
with torch.no_grad():
    for images, labels in testloader:
        images, labels = images.to(device), labels.to(device)
        outputs = model(images)
        _, predicted = torch.max(outputs.data, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

accuracy = 100 * correct / total
print("\n--- Results ---")
print(f"Training Time: {training_duration:.2f} seconds")
print(f"Test Accuracy: {accuracy:.2f}%")
```

## **Question 7: Use a pre-trained VGG16 model (via transfer learning) on a small custom dataset (e.g., flowers or animals). Replace the top layers and fine-tune the model. Include your code and result discussion.**

### **Answer:**

A **pre-trained VGG16 model** can be effectively used for image classification on a small custom dataset using **transfer learning**. Since VGG16 is trained on the large ImageNet dataset, it has already learned rich and generic visual features that can be reused for new but related tasks.

---

### **Methodology**

1. Load the VGG16 model with ImageNet weights and remove the top classification layers.
2. Freeze the convolutional base to retain learned feature representations.
3. Add new fully connected layers suitable for the custom dataset.
4. Fine-tune selected top layers to improve performance on the new data.

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import datasets, models, transforms
import time

# 1. Data Augmentation and Normalization
# VGG16 expects 224x224 images normalized to ImageNet standards
data_transforms = transforms.Compose([
    transforms.Resize(256),
    transforms.CenterCrop(224),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
])

# Loading a subset of Flower dataset (using ImageFolder if you have local files)
# For this snippet, we assume a standard PyTorch Dataset structure
train_dataset = datasets.Flowers102(root='./data', split='train', download=True, transform=data_transforms)
train_loader = torch.utils.data.DataLoader(train_dataset, batch_size=32, shuffle=True)

# 2. Load Pre-trained VGG16
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = models.vgg16(weights=models.VGG16_Weights.IMAGENET1K_V1)

# 3. Freeze Convolutional Layers
for param in model.features.parameters():
    param.requires_grad = False

# 4. Replace the Top Layers (Classifier)
# VGG16 classifier input is 25088 (512 * 7 * 7)
n_inputs = model.classifier[6].in_features
model.classifier[6] = nn.Sequential(
    nn.Linear(n_inputs, 256),
    nn.ReLU(),
    nn.Dropout(0.4),
    nn.Linear(256, 102), # 102 classes for Flowers dataset
    nn.LogSoftmax(dim=1)
)

model = model.to(device)

# 5. Define Loss and Optimizer
# Only the parameters of the classifier are being optimized
criterion = nn.NLLLoss()
optimizer = optim.Adam(model.classifier.parameters(), lr=0.001)

# 6. Training Loop (Condensed)
print(f"Training started on {device}...")
model.train()
start_time = time.time()

for epoch in range(3):  # Small number of epochs for demonstration
    running_loss = 0.0
    for inputs, labels in train_loader:
        inputs, labels = inputs.to(device), labels.to(device)

        optimizer.zero_grad()
        outputs = model(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        running_loss += loss.item()

    print(f"Epoch {epoch+1} - Loss: {running_loss/len(train_loader):.4f}")

total_time = time.time() - start_time
print(f"\nTraining Complete in: {total_time:.2f}s")
```

## **Question 8: Write a program to visualize the filters and feature maps of the first convolutional layer of AlexNet on an example input image.**

### **Answer:**

Visualizing the **filters and feature maps** of the first convolutional layer helps in understanding what kind of low-level features (such as edges, corners, and textures) a convolutional neural network learns from an input image.

Below is an example implementation using **PyTorch** to visualize the feature maps of the **first convolutional layer of AlexNet**.


```python
import torch
import matplotlib.pyplot as plt
from torchvision import models, transforms
from PIL import Image
import requests
from io import BytesIO

# 1. Load Pre-trained AlexNet
model = models.alexnet(weights=models.AlexNet_Weights.IMAGENET1K_V1)
model.eval()

# 2. Get an Example Image (a Dog)
url = "https://raw.githubusercontent.com/pytorch/hub/master/images/dog.jpg"
response = requests.get(url)
img = Image.open(BytesIO(response.content))

# Preprocess for AlexNet (224x224 and ImageNet normalization)
preprocess = transforms.Compose([
    transforms.Resize(256),
    transforms.CenterCrop(224),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
])
input_tensor = preprocess(img).unsqueeze(0)

# 3. Access the First Convolutional Layer
first_conv_layer = model.features[0]

# --- PART A: VISUALIZE FILTERS ---
kernels = first_conv_layer.weight.data.clone().cpu()
# Normalize kernels to [0, 1] for visualization
k_min, k_max = kernels.min(), kernels.max()
kernels = (kernels - k_min) / (k_max - k_min)

fig, axes = plt.subplots(8, 8, figsize=(10, 10))
fig.suptitle("AlexNet: 64 Filters (Weights) of Layer 1", fontsize=16)
for i, ax in enumerate(axes.flat):
    # Convert (C, H, W) -> (H, W, C) for imshow
    ax.imshow(kernels[i].permute(1, 2, 0))
    ax.axis('off')
plt.show()

# --- PART B: VISUALIZE FEATURE MAPS ---
with torch.no_grad():
    # Pass image through only the first layer
    feature_maps = first_conv_layer(input_tensor)

fig, axes = plt.subplots(8, 8, figsize=(10, 10))
fig.suptitle("AlexNet: 64 Feature Maps (Activations) for Input Image", fontsize=16)
for i, ax in enumerate(axes.flat):
    # Map [0, i] because batch size is 1
    ax.imshow(feature_maps[0, i].cpu(), cmap='viridis')
    ax.axis('off')
plt.show()
```

## **Question 9: Train a GoogLeNet (Inception v1) or its variant using a standard dataset like CIFAR-10. Plot the training and validation accuracy over epochs and analyze overfitting or underfitting.**

### **Answer:**

**GoogLeNet (Inception v1)** is a deep convolutional neural network that uses **Inception modules** to efficiently capture features at multiple scales. It achieves high performance with fewer parameters compared to very deep traditional CNNs.

### **Training GoogLeNet on CIFAR-10**

The CIFAR-10 dataset consists of 60,000 color images across 10 classes. A GoogLeNet or its variant can be trained on this dataset to evaluate its learning behavior.


```python
import torch
import torch.nn as nn
import torch.optim as optim
import torchvision
import torchvision.transforms as transforms
import matplotlib.pyplot as plt

# 1. Data Augmentation and Loading
transform_train = transforms.Compose([
    transforms.RandomHorizontalFlip(),
    transforms.RandomCrop(32, padding=4),
    transforms.ToTensor(),
    transforms.Normalize((0.4914, 0.4822, 0.4465), (0.2023, 0.1994, 0.2010)),
])

transform_test = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.4914, 0.4822, 0.4465), (0.2023, 0.1994, 0.2010)),
])

trainset = torchvision.datasets.CIFAR10(root='./data', train=True, download=True, transform=transform_train)
trainloader = torch.utils.data.DataLoader(trainset, batch_size=128, shuffle=True)

testset = torchvision.datasets.CIFAR10(root='./data', train=False, download=True, transform=transform_test)
testloader = torch.utils.data.DataLoader(testset, batch_size=100, shuffle=False)

# 2. Define a simplified Inception Module for CIFAR
class Inception(nn.Module):
    def __init__(self, in_planes, n1x1, n3x3red, n3x3, n5x5red, n5x5, pool_planes):
        super(Inception, self).__init__()
        # 1x1 conv branch
        self.b1 = nn.Sequential(
            nn.Conv2d(in_planes, n1x1, kernel_size=1),
            nn.BatchNorm2d(n1x1), nn.ReLU(True),
        )
        # 1x1 conv -> 3x3 conv branch
        self.b2 = nn.Sequential(
            nn.Conv2d(in_planes, n3x3red, kernel_size=1),
            nn.BatchNorm2d(n3x3red), nn.ReLU(True),
            nn.Conv2d(n3x3red, n3x3, kernel_size=3, padding=1),
            nn.BatchNorm2d(n3x3), nn.ReLU(True),
        )
        # 1x1 conv -> 5x5 conv branch
        self.b3 = nn.Sequential(
            nn.Conv2d(in_planes, n5x5red, kernel_size=1),
            nn.BatchNorm2d(n5x5red), nn.ReLU(True),
            nn.Conv2d(n5x5red, n5x5, kernel_size=3, padding=1), # Using 3x3 for speed on CIFAR
            nn.BatchNorm2d(n5x5), nn.ReLU(True),
        )
        # 3x3 pool -> 1x1 conv branch
        self.b4 = nn.Sequential(
            nn.MaxPool2d(3, stride=1, padding=1),
            nn.Conv2d(in_planes, pool_planes, kernel_size=1),
            nn.BatchNorm2d(pool_planes), nn.ReLU(True),
        )

    def forward(self, x):
        return torch.cat([self.b1(x), self.b2(x), self.b3(x), self.b4(x)], 1)

# 3. Define GoogLeNet
class GoogLeNet(nn.Module):
    def __init__(self):
        super(GoogLeNet, self).__init__()
        self.pre_layers = nn.Sequential(
            nn.Conv2d(3, 192, kernel_size=3, padding=1),
            nn.BatchNorm2d(192), nn.ReLU(True),
        )
        self.a3 = Inception(192,  64,  96, 128, 16, 32, 32)
        self.b3 = Inception(256, 128, 128, 192, 32, 96, 64)
        self.maxpool = nn.MaxPool2d(3, stride=2, padding=1)
        self.a4 = Inception(480, 192,  96, 208, 16, 48, 64)
        self.avgpool = nn.AdaptiveAvgPool2d((1, 1))
        self.linear = nn.Linear(512, 10)

    def forward(self, x):
        out = self.pre_layers(x)
        out = self.a3(out)
        out = self.b3(out)
        out = self.maxpool(out)
        out = self.a4(out)
        out = self.avgpool(out)
        out = out.view(out.size(0), -1)
        out = self.linear(out)
        return out

# 4. Training Loop
device = 'cuda' if torch.cuda.is_available() else 'cpu'
net = GoogLeNet().to(device)
criterion = nn.CrossEntropyLoss()
optimizer = optim.SGD(net.parameters(), lr=0.1, momentum=0.9, weight_decay=5e-4)

history = {'train_acc': [], 'test_acc': []}

for epoch in range(10):
    net.train()
    correct, total = 0, 0
    for inputs, targets in trainloader:
        inputs, targets = inputs.to(device), targets.to(device)
        optimizer.zero_grad()
        outputs = net(inputs)
        loss = criterion(outputs, targets)
        loss.backward()
        optimizer.step()

        _, predicted = outputs.max(1)
        total += targets.size(0)
        correct += predicted.eq(targets).sum().item()

    train_acc = 100.*correct/total
    history['train_acc'].append(train_acc)

    # Validation
    net.eval()
    correct, total = 0, 0
    with torch.no_grad():
        for inputs, targets in testloader:
            inputs, targets = inputs.to(device), targets.to(device)
            outputs = net(inputs)
            _, predicted = outputs.max(1)
            total += targets.size(0)
            correct += predicted.eq(targets).sum().item()
    test_acc = 100.*correct/total
    history['test_acc'].append(test_acc)
    print(f'Epoch {epoch+1}: Train Acc {train_acc:.2f}%, Test Acc {test_acc:.2f}%')

# 5. Plotting
plt.plot(history['train_acc'], label='Train Accuracy')
plt.plot(history['test_acc'], label='Validation Accuracy')
plt.title('GoogLeNet on CIFAR-10')
plt.xlabel('Epochs')
plt.ylabel('Accuracy (%)')
plt.legend()
plt.show()
```
