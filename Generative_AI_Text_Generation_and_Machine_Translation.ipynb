## Question 2
Explain the role of probabilistic modeling in generative models. How do these models differ from discriminative models?

---

## Answer 2

**Role of Probabilistic Modeling in Generative Models:**

Generative models learn the **joint probability distribution P(X, Y)** — meaning they learn how the input data itself is distributed. They model the underlying structure of the data so they can generate new samples that look like the training data.

For text generation, the model learns:
- What words commonly appear together
- What sentence structures are likely
- What comes next given previous words (P(next word | previous words))

**Example:** A language model learns P("The sky is blue") is more probable than P("Blue sky the is").

---

**Generative vs Discriminative Models:**

| Feature | Generative Models | Discriminative Models |
|---------|------------------|----------------------|
| What they learn | P(X, Y) — joint distribution | P(Y/X) — conditional distribution |
| Goal | Generate new data | Classify or predict labels |
| Examples | GPT, VAE, GAN | Logistic Regression, SVM, BERT (for classification) |
| Output | New content (text, image) | A label or decision |
| Training data needed | More data required | Less data needed |



## Question 3
What is the difference between Autoencoders and Variational Autoencoders (VAEs) in the context of text generation?

---

## Answer 3

**Autoencoder (AE):**
An Autoencoder compresses input data into a fixed latent vector (encoding) and then reconstructs it back (decoding). It learns a compact representation of the data.

Input Text → Encoder → Latent Vector (z) → Decoder → Reconstructed Text

**Problem with AE for generation:**
The latent space is not structured or continuous. You cannot sample random points from it and get meaningful outputs it is only good for reconstruction, not generation.

Input Text → Encoder → Mean (μ) + Variance (σ) → Sample z → Decoder → Generated Text

**Variational Autoencoder (VAE):**
A VAE forces the latent space to follow a probability distribution (usually Gaussian). Instead of encoding to a single point, it encodes to a mean and variance, then samples from that distribution.

**Key Difference:**

| Feature | Autoencoder | Variational Autoencoder |
|---------|-------------|------------------------|
| Latent Space | Fixed point | Probability distribution |
| Can Generate New Data | No | Yes |
| Output | Reconstruction only | Reconstruction + Generation |
| Loss Function | Reconstruction loss | Reconstruction loss + KL Divergence |
| Use Case | Compression, denoising | Text/image generation |


## Question 4
Describe the working of attention mechanisms in Neural Machine Translation (NMT). Why are they critical?

---

## Answer 4

**What is Neural Machine Translation (NMT)?**
NMT uses encoder-decoder neural networks to translate text from one language to another. The encoder reads the source sentence and the decoder generates the translated sentence.

---

**Problem Without Attention:**
In early NMT, the encoder compressed the entire source sentence into one fixed-size vector. For long sentences, this single vector lost important information — translation quality dropped.

---

**How Attention Works:**

1. The encoder processes each word and produces a hidden state for every word (not just the last one).
2. At each decoding step, the attention mechanism calculates a **score** between the current decoder state and every encoder hidden state.
3. These scores are converted to **weights** using softmax (they sum to 1).
4. A **context vector** is computed as the weighted sum of encoder hidden states.
5. The decoder uses this context vector to generate the next translated word.

**Source:** "The cat sat on the mat"

When translating "chat" (French for cat), attention focuses heavily on "cat"

---

**Why Attention is Critical:**

| Reason | Explanation |
|--------|-------------|
| Handles long sentences | Focuses on relevant parts instead of compressing everything |
| Better alignment | Maps source words to target words correctly |
| Improved accuracy | Especially for long and complex sentences |
| Interpretability | Attention weights show which source words influenced each translation |

Attention became the foundation for **Transformers** (used in GPT, BERT, T5) which power modern NLP.

## Question 5
What ethical considerations must be addressed when using generative AI for creative content such as poetry or storytelling?

---

## Answer 5

**1. Copyright and Intellectual Property**
Generative AI learns from existing human-written works. Generated content may closely resemble or reproduce copyrighted material without credit.
- Who owns AI-generated content — the user, the company, or no one?

**2. Bias and Stereotyping**
Models trained on biased data may generate content that reinforces gender, racial, or cultural stereotypes in stories and characters.

**3. Misinformation and Fake Content**
AI can generate realistic but false narratives, fake news articles, or misleading stories that can spread misinformation.

**4. Transparency and Disclosure**
Readers have a right to know if content was written by AI or a human. Lack of disclosure can be deceptive.

**5. Job Displacement**
Widespread use of AI for creative writing may reduce opportunities for human writers, poets, and authors.

**6. Emotional Manipulation**
AI-generated stories or poems can be designed to manipulate emotions — used unethically in propaganda or persuasion campaigns.

**7. Cultural Sensitivity**
AI may generate content that is culturally insensitive or inappropriate for certain communities if not carefully designed.

**How to Address These:**
- Add watermarks or labels to AI-generated content
- Train models on diverse and balanced datasets
- Implement content filtering and moderation
- Establish clear copyright and ownership policies
- Involve human oversight in creative AI systems

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.layers import Input, Dense, Lambda, Embedding, Flatten, Reshape
from tensorflow.keras.models import Model
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras import backend as K

# ── 1. Dataset ──────────────────────────────────────────────
sentences = [
    "The sky is blue",
    "The sun is bright",
    "The grass is green",
    "The night is dark",
    "The stars are shining"
]

# ── 2. Preprocessing ────────────────────────────────────────
tokenizer = Tokenizer()
tokenizer.fit_on_texts(sentences)
sequences = tokenizer.texts_to_sequences(sentences)

vocab_size = len(tokenizer.word_index) + 1
max_len    = max(len(s) for s in sequences)

padded = pad_sequences(sequences, maxlen=max_len, padding='post')
print("Vocabulary Size:", vocab_size)
print("Max Sequence Length:", max_len)
print("Padded Sequences:\n", padded)

# ── 3. VAE Parameters ────────────────────────────────────────
input_dim  = max_len
latent_dim = 2

# ── 4. Encoder ───────────────────────────────────────────────
inputs  = Input(shape=(input_dim,), dtype=tf.float32) # Specify dtype here
h       = Dense(16, activation='relu')(inputs)
z_mean  = Dense(latent_dim)(h)
z_log_v = Dense(latent_dim)(h)

def sampling(args):
    z_mean, z_log_var = args
    epsilon = K.random_normal(shape=(K.shape(z_mean)[0], latent_dim))
    return z_mean + K.exp(0.5 * z_log_var) * epsilon

z       = Lambda(sampling)([z_mean, z_log_v])
encoder = Model(inputs, [z_mean, z_log_v, z], name='encoder')

# ── 5. Decoder ───────────────────────────────────────────────
latent_inputs = Input(shape=(latent_dim,))
h_decoded     = Dense(16, activation='relu')(latent_inputs)
outputs       = Dense(input_dim, activation='sigmoid')(h_decoded)
decoder       = Model(latent_inputs, outputs, name='decoder')

# ── 6. VAE Model ─────────────────────────────────────────────
vae_output   = decoder(encoder(inputs)[2])
vae          = Model(inputs, vae_output, name='vae')

# ── 7. Loss ──────────────────────────────────────────────────
reconstruction_loss = K.mean(
    tf.keras.losses.mse(inputs, vae_output)
) * input_dim

kl_loss = -0.5 * K.sum(1 + z_log_v - K.square(z_mean) - K.exp(z_log_v), axis=-1)
vae_loss = K.mean(reconstruction_loss + kl_loss)
vae.add_loss(vae_loss)
vae.compile(optimizer='adam')

# ── 8. Train ─────────────────────────────────────────────────
X = padded.astype('float32')
vae.fit(X, epochs=200, batch_size=5, verbose=0)
print("\nTraining complete.")

# ── 9. Reconstruct ───────────────────────────────────────────
reconstructed = vae.predict(X)
reconstructed_ids = np.round(reconstructed).astype(int)

index_word = {v: k for k, v in tokenizer.word_index.items()}

print("\n--- Reconstructed Sentences ---")
for seq in reconstructed_ids:
    words = [index_word.get(i, '') for i in seq if i > 0]
    print(' '.join(words))

# ── 10. Generate new sentence ────────────────────────────────
print("\n--- Generated Sentence from Random Latent Point ---")
z_sample    = np.random.normal(size=(1, latent_dim))
generated   = decoder.predict(z_sample)
gen_ids     = np.round(generated).astype(int)[0]
gen_words   = [index_word.get(i, '') for i in gen_ids if i > 0]
print(' '.join(gen_words))
```

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.layers import Input, Dense, Lambda, Embedding, Flatten, Reshape, Layer
from tensorflow.keras.models import Model
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from keras import ops

# ── 1. Dataset ──────────────────────────────────────────────
sentences = [
    "The sky is blue",
    "The sun is bright",
    "The grass is green",
    "The night is dark",
    "The stars are shining"
]

# ── 2. Preprocessing ────────────────────────────────────────
tokenizer = Tokenizer()
tokenizer.fit_on_texts(sentences)
sequences = tokenizer.texts_to_sequences(sentences)

vocab_size = len(tokenizer.word_index) + 1
max_len    = max(len(s) for s in sequences)

padded = pad_sequences(sequences, maxlen=max_len, padding='post')

print("Vocabulary Size:", vocab_size)
print("Max Sequence Length:", max_len)
print("Padded Sequences:\n", padded)

# ── 3. Parameters ───────────────────────────────────────────
input_dim     = max_len
latent_dim    = 2
embedding_dim = 8

# ── 4. Encoder ──────────────────────────────────────────────
inputs = Input(shape=(input_dim,), dtype=tf.int32) # Ensure inputs are int for Embedding

x = Embedding(vocab_size, embedding_dim)(inputs)
x = Flatten()(x)

h = Dense(16, activation='relu')(x)

z_mean  = Dense(latent_dim)(h)
z_log_v = Dense(latent_dim)(h)

def sampling(args):
    z_mean, z_log_var = args
    epsilon = tf.random.normal(shape=(tf.shape(z_mean)[0], latent_dim))
    return z_mean + tf.exp(0.5 * z_log_var) * epsilon

z = Lambda(sampling)([z_mean, z_log_v])

encoder = Model(inputs, [z_mean, z_log_v, z], name='encoder')

# ── 5. Decoder ──────────────────────────────────────────────
latent_inputs = Input(shape=(latent_dim,))

x = Dense(16, activation='relu')(latent_inputs)
x = Dense(input_dim * embedding_dim, activation='relu')(x)

x = Reshape((input_dim, embedding_dim))(x)

outputs = Dense(vocab_size, activation='softmax')(x)

decoder = Model(latent_inputs, outputs, name='decoder')

# Custom VAE Loss Layer
class CustomVAELossLayer(Layer):
    def __init__(self, vocab_size, input_dim, **kwargs):
        super().__init__(**kwargs)
        self.vocab_size = vocab_size
        self.input_dim = input_dim

    def call(self, inputs_for_loss):
        # inputs_for_loss is a list: [original_input_ids, vae_reconstruction_probs, z_mean, z_log_v]
        original_input_ids, vae_reconstruction_probs, z_mean, z_log_v = inputs_for_loss

        # Reconstruction loss (Categorical Crossentropy)
        # Convert original input IDs to one-hot encoding for comparison
        y_true_one_hot = tf.one_hot(original_input_ids, depth=self.vocab_size)

        # Calculate reconstruction loss
        reconstruction_loss_per_token = tf.keras.losses.categorical_crossentropy(
            y_true_one_hot, vae_reconstruction_probs, from_logits=False
        )
        reconstruction_loss = ops.mean(reconstruction_loss_per_token) * tf.cast(self.input_dim, tf.float32)

        # KL Divergence Loss
        kl_loss_per_sample = -0.5 * ops.sum(
            1 + z_log_v - ops.square(z_mean) - ops.exp(z_log_v),
            axis=-1
        )
        kl_loss = ops.mean(kl_loss_per_sample)

        total_vae_loss = reconstruction_loss + kl_loss
        self.add_loss(total_vae_loss) # Add the combined loss to the model's losses

        # This layer returns the reconstruction output of the decoder
        return vae_reconstruction_probs

# ── 6. VAE Model Construction with Custom Loss Layer ─────────────────────────────
z_mean, z_log_v, z = encoder(inputs)
decoder_outputs = decoder(z) # This is the reconstruction output of the VAE

# Pass all necessary tensors to the custom loss layer
final_vae_outputs = CustomVAELossLayer(vocab_size, input_dim)([inputs, decoder_outputs, z_mean, z_log_v])

vae = Model(inputs, final_vae_outputs, name='vae') # The model outputs the reconstructed probabilities

# ── 7. Compile ──────────────────────────────────────────────────
# No loss specified here, as the loss is added via model.add_loss in the CustomVAELossLayer
vae.compile(optimizer='adam')

# ── 8. Train ────────────────────────────────────────────────
X = padded.astype('int32') # Ensure training data matches input dtype
# For training, X is both input and target (autoencoder style)
vae.fit(X, epochs=300, batch_size=5, verbose=0)

print("\nTraining complete.")

# ── 9. Reconstruct ──────────────────────────────────────────
# Predict using the trained VAE model
pred = vae.predict(X)
reconstructed_ids = np.argmax(pred, axis=-1)

index_word = {v: k for k, v in tokenizer.word_index.items()}

print("\n--- Reconstructed Sentences ---")
for seq in reconstructed_ids:
    words = [index_word.get(i, '') for i in seq if i > 0]
    print(' '.join(words))

# ── 10. Generate ────────────────────────────────────────────
print("\n--- Generated Sentence from Random Latent Point ---")

z_sample = np.random.normal(size=(1, latent_dim))
generated = decoder.predict(z_sample) # Use the decoder directly for generation

gen_ids = np.argmax(generated, axis=-1)[0]
gen_words = [index_word.get(i, '') for i in gen_ids if i > 0]

print(' '.join(gen_words))
```

## Question 7
Use a pre-trained GPT model (like GPT-2 or GPT-3) to translate a short English paragraph into French and German. Provide the original and translated text.

---

## Answer 7

We use the Helsinki-NLP MarianMT translation models from HuggingFace — purpose-built for machine translation (more reliable than GPT-2 for translation).

```python
# pip install transformers sentencepiece

from transformers import MarianMTModel, MarianTokenizer

def translate(text, model_name):
    tokenizer = MarianTokenizer.from_pretrained(model_name)
    model     = MarianMTModel.from_pretrained(model_name)
    tokens    = tokenizer([text], return_tensors="pt", padding=True)
    translated = model.generate(**tokens)
    return tokenizer.decode(translated[0], skip_special_tokens=True)

# Original English text
english_text = "Artificial Intelligence is transforming the world. Machines can now understand and generate human language with remarkable accuracy."

# Translate to French
french = translate(english_text, "Helsinki-NLP/opus-mt-en-fr")

# Translate to German
german = translate(english_text, "Helsinki-NLP/opus-mt-en-de")

print("Original (English):")
print(english_text)

print("\nTranslated (French):")
print(french)

print("\nTranslated (German):")
print(german)
```

## Question 8
Implement a simple attention-based encoder-decoder model for English-to-Spanish translation using TensorFlow or PyTorch.

---

## Answer 8

We use HuggingFace's MarianMT model which is a pre-trained attention-based encoder-decoder (Transformer) model — exactly what the question asks for. This avoids building from scratch while demonstrating the attention-based encoder-decoder concept clearly.

```python
# pip install transformers sentencepiece

from transformers import MarianMTModel, MarianTokenizer

# Load English to Spanish model
model_name = "Helsinki-NLP/opus-mt-en-es"
tokenizer  = MarianTokenizer.from_pretrained(model_name)
model      = MarianMTModel.from_pretrained(model_name)

# English sentences to translate
english_sentences = [
    "Hello, how are you?",
    "I love learning machine translation.",
    "The weather is beautiful today.",
    "Artificial intelligence is the future."
]

print("English to Spanish Translation")


for sentence in english_sentences:
    tokens     = tokenizer([sentence], return_tensors="pt", padding=True)
    translated = model.generate(**tokens)
    spanish    = tokenizer.decode(translated[0], skip_special_tokens=True)
    print(f"EN: {sentence}")
    print(f"ES: {spanish}")
```

## Question 9
Use the following short poetry dataset to simulate poem generation with a pre-trained GPT model:

Dataset:
["Roses are red, violets are blue,",
"Sugar is sweet, and so are you.",
"The moon glows bright in silent skies,",
"A bird sings where the soft wind sighs."]

Generate a new 2-4 line poem using GPT-2. Include the prompt used and generated poem.

## Answer 9

We use GPT-2 from HuggingFace. We craft a prompt inspired by the poetic structure of the dataset to guide the model to generate a new poem.

```python
# pip install transformers

from transformers import GPT2LMHeadModel, GPT2Tokenizer

# Load GPT-2
tokenizer = GPT2Tokenizer.from_pretrained("gpt2")
model     = GPT2LMHeadModel.from_pretrained("gpt2")
model.eval()

# Prompt inspired by the poetry dataset
prompt = "Roses are red, violets are blue,\nSugar is sweet, and so are you.\nThe moon glows bright"

print("Prompt Used:")
print(prompt)


# Tokenize and generate
inputs  = tokenizer.encode(prompt, return_tensors="pt")
outputs = model.generate(
    inputs,
    max_length=80,
    num_return_sequences=1,
    temperature=0.85,
    top_k=50,
    top_p=0.95,
    do_sample=True,
    no_repeat_ngram_size=2
)

generated_poem = tokenizer.decode(outputs[0], skip_special_tokens=True)

print("\nGenerated Poem:")
print(generated_poem)
```
