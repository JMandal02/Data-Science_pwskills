# **Question 1: What is a Vector Database (VectorDB) and how is it different from traditional databases?**

**Answer:**

A Vector Database (VectorDB) is a specialized database designed to store, index, and query
high-dimensional vector embeddings. These embeddings are numerical representations of unstructured
data (text, images, audio) generated by ML/AI models.

**Key differences from traditional databases:**

**Data Storage:**
  Traditional DBs store structured data in rows and columns (tables).
  VectorDBs store float vectors (embeddings) of high dimensions (e.g., 384-d, 768-d arrays).

**Query Type:**
  Traditional DBs use exact match queries (SQL WHERE, JOIN).
  VectorDBs use Approximate Nearest Neighbor (ANN) search to find semantically similar vectors.

**Indexing:**
  Traditional DBs use B-tree or hash indexes.
  VectorDBs use specialized indexes like HNSW, IVF, LSH optimized for similarity search.

**Use Case:**
  Traditional DBs are used for structured business data (inventory, transactions).
  VectorDBs are used for semantic search, recommendation engines, RAG pipelines, and chatbots.

**Schema:**
  Traditional DBs require a fixed schema.
  VectorDBs are schema-flexible and pair vectors with optional metadata.

**Examples of VectorDBs:** ChromaDB, Pinecone, Weaviate, Milvus, FAISS, Qdrant.

# **Question 2: Explain the various types of VectorDBs available and describe their suitability for different use cases.**


**Answer:**

**1. Purpose-Built VectorDBs (e.g., Pinecone, Weaviate, Qdrant, Milvus)**
   - Built from scratch specifically for vector operations.
   - Support advanced filtering, hybrid search, and large-scale indexing.
   - Best for: Production-grade AI apps, enterprise semantic search, large-scale RAG systems.

**2. Lightweight / Embedded VectorDBs (e.g., ChromaDB, FAISS)**
   - Run in-process, no separate server needed.
   - Easy to install with pip, minimal configuration.
   - Best for: Prototyping, research projects, small to medium scale applications,
     local LLM pipelines.

**3. Vector Extensions on Traditional DBs (e.g., pgvector for PostgreSQL, Redis with vector module)**
   - Adds vector search capability to existing relational or key-value databases.
   - Best for: Teams already using relational DBs who want to add semantic search
     without a new system.

**4. Cloud-Native / Managed VectorDBs (e.g., Pinecone Cloud, Zilliz Cloud)**
   - Fully managed, auto-scaling, no infrastructure maintenance.
   - Best for: Enterprise deployments where uptime, scalability, and SLA matter.

**5. In-Memory VectorDBs (e.g., FAISS, Annoy)**
   - Pure in-memory, ultra-fast similarity search.
   - No persistence by default.
   - Best for: Real-time recommendation systems, short-lived search tasks,
     batch similarity computations.



# **Question 3: Why is Chroma DB considered important in the context of AI/ML projects? Describe its key features.**

**Answer:**

ChromaDB is an open-source, lightweight, in-process vector database designed specifically
for AI and LLM-powered applications. It has become the most popular choice for building
RAG pipelines and semantic search systems quickly.

**Why it is important:**
- Zero infrastructure overhead: runs embedded in Python, no server setup needed.
- Fastest path from idea to working prototype for LLM applications.
- Native integration with LangChain, LlamaIndex, and HuggingFace.
- Free and open source with active community support.

**Key Features:**

**1. Simple Python API**
   - pip install chromadb is all it takes to get started.
   - Easy methods: create_collection(), add(), query().

**2. Multiple Storage Modes**
   - In-memory (ephemeral): for quick testing.
   - Persistent (on-disk): for saving embeddings across sessions.

**3. Built-in Embedding Functions**
   - Supports SentenceTransformers, OpenAI, Cohere embeddings out of the box.

**4. Metadata Filtering**
   - Filter search results by metadata fields alongside vector similarity.

**5. Integration with LLM Frameworks**
   - Direct integration with LangChain as a vectorstore.
   - Works well with LlamaIndex retrieval pipelines.

**6. Scalability**
   - Suitable for small-scale local use and medium-scale production deployments.

# **Question 6: Install and set up Chroma DB, and insert sample vector data for semantic search.**

```python
# Install required libraries
!pip install chromadb sentence-transformers -q
```

```python
import chromadb
from chromadb.utils import embedding_functions

# Use SentenceTransformer embedding function (free, no API key needed)
embedding_fn = embedding_functions.SentenceTransformerEmbeddingFunction(
    model_name="all-MiniLM-L6-v2"
)

# Create in-memory ChromaDB client and a collection
client = chromadb.Client()
collection = client.create_collection(
    name="sample_docs",
    embedding_function=embedding_fn
)

# Sample documents to insert
documents = [
    "Artificial intelligence is transforming the healthcare industry.",
    "Machine learning models require large amounts of labelled data to train.",
    "Deep learning is a subset of machine learning based on neural networks.",
    "Natural language processing enables computers to understand human text.",
    "Vector databases are optimized for semantic similarity search."
]

ids = [f"doc{i}" for i in range(len(documents))]

# Insert documents into the collection
collection.add(documents=documents, ids=ids)
print(f"Inserted {len(documents)} documents into ChromaDB.\n")

# Perform semantic search
query = "How does AI help in medicine?"
results = collection.query(query_texts=[query], n_results=2)

print(f"Query: {query}")
print("\nTop 2 Semantically Similar Results:")
for i, (doc, dist) in enumerate(zip(results['documents'][0], results['distances'][0])):
    print(f"Rank {i+1}: {doc}")
    print(f"Distance: {dist:.4f}")
```

# **Question 7: Demonstrate how to download and fine-tune a Hugging Face model for a text generation task.**


```python
# Install required libraries
!pip install transformers datasets accelerate -q
```

```python
import torch
from transformers import (
    GPT2LMHeadModel,
    GPT2Tokenizer,
    Trainer,
    TrainingArguments,
    DataCollatorForLanguageModeling,
    pipeline
)
from datasets import Dataset

# Step 1: Load Model
print("Step 1: Loading GPT-2 model and tokenizer...")
model_name = "gpt2"

tokenizer = GPT2Tokenizer.from_pretrained(model_name)
tokenizer.pad_token = tokenizer.eos_token  # GPT2 has no pad token

model = GPT2LMHeadModel.from_pretrained(model_name)
print("Model loaded successfully.\n")



# Step 2: Prepare Dataset
print("Step 2: Preparing dataset...")

train_texts = [
    "Artificial intelligence is the simulation of human intelligence by machines.",
    "Python is the most popular language for machine learning and data science.",
    "Deep learning uses neural networks with many layers to learn patterns.",
    "Natural language processing helps machines understand and generate text.",
    "Transfer learning allows models to apply knowledge from one task to another.",
    "Supervised learning trains models using labelled input-output pairs.",
    "Reinforcement learning trains agents by rewarding correct actions.",
    "Embeddings represent words or sentences as dense numerical vectors.",
]

dataset = Dataset.from_dict({"text": train_texts})


def tokenize_function(examples):
    return tokenizer(
        examples["text"],
        truncation=True,
        padding="max_length",
        max_length=64
    )


tokenized_dataset = dataset.map(tokenize_function, batched=True)

# IMPORTANT: Add labels for GPT-2 training
tokenized_dataset = tokenized_dataset.map(
    lambda x: {"labels": x["input_ids"]},
    batched=True
)

tokenized_dataset = tokenized_dataset.remove_columns(["text"])

print(f"Dataset ready. Total samples: {len(tokenized_dataset)}\n")



# Step 3: Training Setup
print("Step 3: Fine-tuning GPT-2...")

training_args = TrainingArguments(
    output_dir="./gpt2-finetuned",
    num_train_epochs=3,
    per_device_train_batch_size=2,
    logging_steps=5,
    save_steps=50,
    report_to="none"  # disable wandb
)

data_collator = DataCollatorForLanguageModeling(
    tokenizer=tokenizer,
    mlm=False  # GPT-2 uses causal LM
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_dataset,
    data_collator=data_collator
)

trainer.train()
print("Fine-tuning complete.\n")


# Step 4: Text Generation
print("Step 4: Testing model...")

generator = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    device=0 if torch.cuda.is_available() else -1
)

prompts = [
    "Artificial intelligence",
    "Deep learning is",
    "Python is used for"
]

for prompt in prompts:
    output = generator(
        prompt,
        max_new_tokens=30,
        num_return_sequences=1,
        pad_token_id=tokenizer.eos_token_id
    )
    print(f"\nPrompt : {prompt}")
    print(f"Output : {output[0]['generated_text']}")
```

# **Question 8: Create a custom LLM using Ollama and Llama2, and run it locally for basic text prompts.**

```python
# Install required libraries
!pip install transformers -q
```

```python
import torch
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

print("Loading custom LLM (google/flan-t5-base)...")

model_name = "google/flan-t5-base"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(
    model_name,
    device_map="auto" if torch.cuda.is_available() else None
)

print("Custom LLM loaded successfully.\n")


def run_llm(prompt, max_new_tokens=150):
    inputs = tokenizer(prompt, return_tensors="pt")

    if torch.cuda.is_available():
        inputs = {k: v.to(model.device) for k, v in inputs.items()}

    outputs = model.generate(
        **inputs,
        max_new_tokens=max_new_tokens,
        do_sample=False
    )

    response = tokenizer.decode(outputs[0], skip_special_tokens=True)
    return response


prompts = [
    "What is artificial intelligence?",
    "Explain the difference between supervised and unsupervised learning.",
    "What is a large language model?",
    "List three applications of natural language processing.",
    "What is the role of embeddings in NLP?"
]

print("CUSTOM LLM - TEXT PROMPT RESPONSES")
for prompt in prompts:
    response = run_llm(prompt)
    print(f"\nPrompt  : {prompt}")
    print(f"Response: {response}")
```

# **Question 9: Implement a basic RAG (Retrieval-Augmented Generation) system using Ollama with Llama3.**

```python
# Install required libraries
!pip install chromadb sentence-transformers transformers -q
```

```python
import torch
import chromadb
from chromadb.utils import embedding_functions
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

# ── STEP 1: Knowledge Base ─────────────────────────────────────
print("Step 1: Setting up ChromaDB knowledge base...")

ef = embedding_functions.SentenceTransformerEmbeddingFunction(
    model_name="all-MiniLM-L6-v2"
)

client = chromadb.Client()
collection = client.create_collection(
    "rag_knowledge_base",
    embedding_function=ef
)

documents = [
    "Python was created by Guido van Rossum and first released in 1991.",
    "LangChain is a framework for building applications powered by large language models.",
    "RAG stands for Retrieval-Augmented Generation. It combines a retriever and a generator.",
    "ChromaDB is an open-source vector database used for storing and querying embeddings.",
    "The Transformer architecture was introduced in the 2017 paper Attention is All You Need.",
    "Fine-tuning adapts a pre-trained model to a specific task using a smaller dataset.",
    "FAISS is a library developed by Meta for efficient similarity search on dense vectors.",
    "Embeddings are dense numerical vector representations of text or other data."
]

collection.add(
    documents=documents,
    ids=[f"doc_{i}" for i in range(len(documents))]
)

print(f"Knowledge base ready with {len(documents)} documents.\n")


# ── STEP 2: Load FLAN-T5 Properly ───────────────────────────────
print("Step 2: Loading LLM (flan-t5-base)...")

model_name = "google/flan-t5-base"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(
    model_name,
    device_map="auto" if torch.cuda.is_available() else None
)

print("LLM loaded.\n")


# ── STEP 3: RAG Pipeline ────────────────────────────────────────
def generate_answer(prompt, max_new_tokens=120):
    inputs = tokenizer(prompt, return_tensors="pt")

    if torch.cuda.is_available():
        inputs = {k: v.to(model.device) for k, v in inputs.items()}

    outputs = model.generate(
        **inputs,
        max_new_tokens=max_new_tokens,
        do_sample=True,
        temperature=0.7
    )

    return tokenizer.decode(outputs[0], skip_special_tokens=True)


def rag_pipeline(question, n_results=2):
    # Retrieval
    results = collection.query(query_texts=[question], n_results=n_results)
    retrieved_docs = results['documents'][0]

    context = " ".join(retrieved_docs)

    # IMPORTANT: Instruction-style prompt
    prompt = f"""
Answer the question based on the context below.

Context:
{context}

Question:
{question}

Answer:
"""

    answer = generate_answer(prompt)

    return {
        "question": question,
        "retrieved_context": retrieved_docs,
        "answer": answer
    }


# ── STEP 4: Test ────────────────────────────────────────────────
test_questions = [
    "Who created Python and when?",
    "What is RAG in machine learning?",
    "What is ChromaDB used for?",
    "What is fine-tuning in deep learning?"
]

print("RAG SYSTEM OUTPUT")
for q in test_questions:
    result = rag_pipeline(q)
    print(f"\nQuestion : {result['question']}")
    print(f"Retrieved: {result['retrieved_context'][0][:80]}...")
    print(f"Answer: {result['answer']}")
```

# **Question 10: A health-tech startup wants to build a chatbot that can answer user queries based on medical research articles.Propose and explain a solution using Hugging Face models for understanding,VectorDB for retrieval,and Ollama for generation.**


### **RAG System Architecture**

**Components:**
- Embedding Model → `all-MiniLM-L6-v2` (Hugging Face)
- Vector Database → ChromaDB
- LLM Generator → `flan-t5-base` (Hugging Face)

**Pipeline Flow:**
User Query  
→ Convert to Embedding  
→ Retrieve Relevant Docs (ChromaDB)  
→ Augment Prompt (Context + Question)  
→ Generate Answer (LLM)


```python
# Install required libraries
!pip install chromadb sentence-transformers transformers -q
```

```python
import torch
import chromadb
from chromadb.utils import embedding_functions
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

# ── COMPONENT 1: VectorDB Setup ─────────────────────────────────────────────
print("Setting up medical knowledge base in ChromaDB...")

ef = embedding_functions.SentenceTransformerEmbeddingFunction(
    model_name="all-MiniLM-L6-v2"
)

client = chromadb.Client()
medical_db = client.get_or_create_collection(
    name="medical_research",
    embedding_function=ef
)

medical_articles = [
    "Diabetes mellitus is a chronic metabolic disease characterized by elevated blood sugar. "
    "Type 1 diabetes is autoimmune; Type 2 is linked to insulin resistance. "
    "Treatment includes insulin therapy, metformin, and dietary modification.",

    "Hypertension, or high blood pressure, is a major risk factor for cardiovascular disease. "
    "It is managed using ACE inhibitors, beta-blockers, and lifestyle changes "
    "such as reduced sodium intake and regular exercise.",

    "COVID-19 is caused by the SARS-CoV-2 virus. Common symptoms include fever, dry cough, "
    "fatigue, and loss of taste or smell. Vaccines including mRNA-based ones have been "
    "developed to prevent severe illness and hospitalization.",

    "Antibiotic resistance occurs when bacteria evolve to withstand antibiotic drugs. "
    "It is accelerated by overuse and misuse of antibiotics in humans and livestock. "
    "WHO considers it one of the greatest global health threats.",

    "Regular physical activity of at least 150 minutes per week reduces the risk of "
    "cardiovascular disease, type 2 diabetes, obesity, and certain cancers. "
    "Exercise also improves mental health and cognitive function.",

    "Depression is a common mental health disorder characterized by persistent sadness, "
    "loss of interest, and fatigue. It is treated with cognitive behavioural therapy, "
    "antidepressant medications, and lifestyle interventions.",

    "Cancer is caused by uncontrolled cell growth due to DNA mutations. Common types include "
    "lung, breast, and colorectal cancer. Treatments include chemotherapy, radiation, surgery, "
    "and targeted immunotherapy."
]

ids = [f"article_{i}" for i in range(len(medical_articles))]
medical_db.add(documents=medical_articles, ids=ids)

print(f"Loaded {len(medical_articles)} medical articles.\n")


# ── COMPONENT 2: Load FLAN-T5 (CORRECT WAY) ─────────────────────────────────
print("Loading LLM (flan-t5-base)...")

model_name = "google/flan-t5-base"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(
    model_name,
    device_map="auto" if torch.cuda.is_available() else None
)

print("LLM ready.\n")


# ── COMPONENT 3: Medical Chatbot (RAG Pipeline) ─────────────────────────────
def medical_chatbot(user_query, n_results=2, max_tokens=180):

    # Step 1: Retrieve context
    results = medical_db.query(query_texts=[user_query], n_results=n_results)
    context_docs = results['documents'][0]
    context = " ".join(context_docs)

    # Step 2: Build instruction-style prompt (VERY IMPORTANT)
    prompt = f"""
Answer the question based only on the medical context below.

Context:
{context}

Question:
{user_query}

Answer:
"""

    # Step 3: Generate answer (FIXED)
    inputs = tokenizer(prompt, return_tensors="pt")

    if torch.cuda.is_available():
        inputs = {k: v.to(model.device) for k, v in inputs.items()}

    outputs = model.generate(
        **inputs,
        max_new_tokens=max_tokens,
        do_sample=True,
        temperature=0.7
    )

    answer = tokenizer.decode(outputs[0], skip_special_tokens=True)

    return {
        "query": user_query,
        "context_used": context_docs,
        "answer": answer
    }


# ── COMPONENT 4: Demo ───────────────────────────────────────────────────────
test_queries = [
    "What is diabetes and how is it treated?",
    "How can high blood pressure be managed?",
    "What are the symptoms of COVID-19?",
    "Why is antibiotic resistance dangerous?",
    "How does exercise affect health?"
]

print("MEDICAL CHATBOT DEMO")

for q in test_queries:
    result = medical_chatbot(q)
    print(f"\nUser Query   : {result['query']}")
    print(f"Context Used : {result['context_used'][0][:80]}...")
    print(f"Chatbot Reply: {result['answer']}")


# ── SYSTEM SUMMARY ──────────────────────────────────────────────────────────
print("\nSystem Architecture Summary:")
print("Embedding Model : all-MiniLM-L6-v2 (HuggingFace)")
print("Vector Database : ChromaDB")
print("LLM Generator   : google/flan-t5-base")
print("Pattern         : RAG (Retrieval-Augmented Generation)")
```
