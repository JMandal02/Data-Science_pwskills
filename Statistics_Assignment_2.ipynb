## Question 1: What is hypothesis testing in statistics?

**Answer:**

Hypothesis testing is a statistical method used to make decisions or inferences about population parameters based on sample data.  
It provides a framework to test assumptions (hypotheses) by comparing observed results with expected outcomes under a certain assumption.

**Steps:**
1. State the null and alternative hypothesis  
2. Choose significance level (α)  
3. Select a test statistic  
4. Compute test statistic and p-value  
5. Compare p-value with α to accept or reject the null hypothesis  

---

## Question 2: What is the null hypothesis, and how does it differ from the alternative hypothesis?

**Answer:**

- **Null Hypothesis (H₀):** A statement that there is no effect or no difference. Example: H₀: μ = 50  
- **Alternative Hypothesis (H₁):** A statement that contradicts H₀, suggesting a significant effect. Example: H₁: μ ≠ 50  

**Difference:**  
- H₀ → "no effect"  
- H₁ → research claim  

---

## Question 3: Explain the significance level in hypothesis testing and its role.

**Answer:**

- **Significance Level (α):** Probability of rejecting H₀ when it is true (Type I error).  
- Common values: 0.05 or 0.01.  

**Decision Rule:**  
- If p-value ≤ α → reject H₀  
- If p-value > α → fail to reject H₀  

---

## Question 4: What are Type I and Type II errors? Give examples.

**Answer:**

- **Type I Error (α):** Rejecting a true null hypothesis  
  - Example: A medical test says a healthy person has a disease.  
- **Type II Error (β):** Failing to reject a false null hypothesis  
  - Example: A medical test fails to detect a disease when the person actually has it.  

---

## Question 5: Difference between Z-test and T-test

**Answer:**

- **Z-test**  
  - Used when population variance is known or sample size is large (n ≥ 30).  
  - Based on standard normal distribution.  

- **T-test**  
  - Used when population variance is unknown and sample size is small (n < 30).  
  - Based on t-distribution (heavier tails).  

**When to use:**  
- Z-test → large samples / known variance  
- T-test → small samples / unknown variance

```python
# Question 7: Hypothesis testing using Z-statistics
import numpy as np
from scipy import stats

sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

sample_mean = np.mean(sample_data)
pop_mean = 50
sample_std = np.std(sample_data, ddof=1)
n = len(sample_data)

z_stat = (sample_mean - pop_mean) / (sample_std / np.sqrt(n))
p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))

print("Z-statistic:", z_stat)
print("p-value:", p_value)
```

```python
# Question 8: Normal distribution simulation + 95% Confidence Interval
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

data = np.random.normal(loc=50, scale=5, size=1000)
mean = np.mean(data)
std = np.std(data, ddof=1)
conf_interval = stats.norm.interval(0.95, loc=mean, scale=std/np.sqrt(len(data)))

print("Mean:", mean)
print("95% Confidence Interval:", conf_interval)

plt.hist(data, bins=30, edgecolor='black', density=True)
plt.axvline(mean, color='red', linestyle='dashed', linewidth=2, label=f"Mean={mean:.2f}")
plt.title("Normal Distribution with 95% CI")
plt.legend()
plt.show()
```

```python
# Question 9: Z-scores and histogram
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

def z_score_plot(data):
    z_scores = stats.zscore(data)
    plt.hist(z_scores, bins=20, edgecolor='black', density=True)
    plt.title("Histogram of Z-scores")
    plt.xlabel("Z-score")
    plt.ylabel("Frequency")
    plt.show()
    return z_scores

data = np.random.normal(50, 5, 1000)
z_scores = z_score_plot(data)
print("First 10 Z-scores:", z_scores[:10])
```
