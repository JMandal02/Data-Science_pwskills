## Question 1: What are Large Language Models (LLMs) and how do they function?

**Answer:**

Large Language Models (LLMs) are deep learning models trained on massive text corpora to understand and generate human-like language. They are based on the **Transformer architecture** introduced in *"Attention is All You Need"* (Vaswani et al., 2017).

### What are LLMs?
- LLMs are AI models with billions of parameters trained on diverse internet text, books, code, etc.
- Examples: **GPT-4** (OpenAI), **Gemini** (Google), **LLaMA** (Meta), **Claude** (Anthropic), **Mistral**
- They perform tasks like Q&A, summarization, translation, code generation — without task-specific retraining

### How do LLMs Function?

**1. Tokenization:**
- Input text is split into tokens using tokenizers like BPE (Byte Pair Encoding)
- Each token is mapped to a numeric ID and then to a dense embedding vector

**2. Transformer Architecture:**
- Core mechanism is **Self-Attention** — allows the model to weigh relevance of every token against others
- **Encoder** understands context (BERT-style); **Decoder** generates output autoregressively (GPT-style)

**3. Pre-Training:**
- Trained using **Causal Language Modeling** — predict the next token given previous tokens
- Uses trillions of tokens and massive GPU/TPU compute

**4. Fine-Tuning & RLHF:**
- Fine-tuned on instruction-following datasets after pre-training
- **RLHF (Reinforcement Learning from Human Feedback)** aligns the model to be helpful and safe

**5. Inference:**
- Generates text token-by-token using greedy decoding, beam search, or temperature sampling

## Question 3: What are the key advantages and limitations of using LLMs in real-world applications?

**Answer:**

### Key Advantages

**1. Versatility**
- A single LLM can handle Q&A, summarization, translation, code generation, sentiment analysis, and more without retraining
- Reduces the need to build and maintain multiple specialized models

**2. Natural Language Interface**
- Users interact using plain English (or any language), eliminating the need for structured inputs
- Makes AI accessible to non-technical users

**3. Few-Shot and Zero-Shot Learning**
- LLMs generalize to new tasks with just a few examples in the prompt or even no examples at all
- Drastically reduces labeled data requirements

**4. Contextual Understanding**
- LLMs understand nuance, sarcasm, ambiguity, and long-range dependencies in text
- Far superior to traditional keyword-matching or bag-of-words models

**5. Rapid Development**
- Building LLM-powered applications using frameworks like LangChain takes hours instead of months
- Pre-trained models skip the expensive training phase entirely

**6. Multilingual Capability**
- Modern LLMs natively support dozens of languages without separate models

---

### Key Limitations

**1. Hallucination**
- LLMs confidently generate factually incorrect information
- Critical in domains like healthcare, law, and finance where accuracy is mandatory

**2. Knowledge Cutoff**
- LLMs have a training cutoff date and are unaware of recent events unless augmented with tools like RAG or web search

**3. High Computational Cost**
- Running large models (GPT-4, Claude) requires expensive API calls
- Self-hosting requires significant GPU infrastructure

**4. Context Window Limitations**
- Most LLMs have a limited context window (e.g., 4K–200K tokens)
- Very long documents must be chunked, potentially losing cross-document context

**5. Bias and Toxicity**
- Trained on internet data, LLMs can reflect societal biases and generate harmful content
- Requires careful filtering and alignment techniques

**6. Lack of True Reasoning**
- LLMs perform pattern matching, not genuine logical reasoning
- Complex multi-step reasoning tasks remain challenging

**7. Data Privacy**
- Sending sensitive data to third-party LLM APIs raises GDPR and confidentiality concerns

## Question 5: Compare and contrast Langchain and LlamaIndex. What unique problems does each solve?

**Answer:**

### Overview

| Feature | LangChain | LlamaIndex |
|---|---|---|
| Primary Focus | Building LLM-powered applications & agents | Data indexing & retrieval (RAG) |
| Core Abstraction | Chains, Agents, Tools | Index, Query Engine, Retriever |
| Best For | Complex multi-step workflows, agents | Document Q&A, search over private data |
| Learning Curve | Moderate to High | Low to Moderate |
| Community | Very large | Large and growing |

---

### LangChain

**What it solves:**
LangChain solves the problem of **orchestrating complex LLM workflows** — chaining multiple LLM calls, tools, APIs, and memory together into intelligent pipelines and agents.

**Key Features:**
- **Chains** — Sequential pipelines of LLM calls and transformations
- **Agents** — LLMs that dynamically decide which tools to call (ReAct, OpenAI Functions)
- **Memory** — Conversation history management (short-term and long-term)
- **Tools & Integrations** — 100+ integrations (APIs, databases, search engines)
- **Prompt Templates** — Reusable, parameterized prompts

**Unique Problems Solved:**
- Building multi-step reasoning pipelines
- Creating autonomous AI agents with tool use
- Managing conversational memory across sessions
- Connecting LLMs to external APIs and databases

**Example Use Case:** A customer service agent that retrieves account info from a database, checks order status via API, and drafts an email — all in one pipeline.

---

### LlamaIndex (formerly GPT Index)

**What it solves:**
LlamaIndex solves the problem of **connecting LLMs to your private data** through efficient indexing and retrieval — the core of Retrieval-Augmented Generation (RAG).

**Key Features:**
- **Data Connectors** — Load data from PDFs, Notion, Slack, SQL, Google Drive, etc.
- **Index Types** — Vector Store Index, Summary Index, Knowledge Graph Index
- **Query Engine** — Natural language queries over indexed documents
- **Node Parsers** — Smart chunking and parsing of documents
- **Response Synthesizers** — Combine retrieved chunks into coherent answers

**Unique Problems Solved:**
- Indexing large private document collections for semantic search
- Building RAG pipelines over custom knowledge bases
- Efficient chunking and retrieval strategy optimization
- Handling structured + unstructured data together

**Example Use Case:** A legal firm indexing 10,000 case documents and querying them with natural language — "Find all cases where the defendant was acquitted due to insufficient evidence."

---

### Key Differences

**1. Purpose:**
- LangChain = Application orchestration framework
- LlamaIndex = Data framework for LLM applications

**2. Strengths:**
- LangChain excels at **agent workflows and tool use**
- LlamaIndex excels at **document retrieval and RAG quality**

**3. Complementary Nature:**
- They are often **used together** — LlamaIndex handles the retrieval layer while LangChain handles the agent/orchestration layer


## Question 6: Implement a basic Langchain pipeline using OpenAI's LLM to answer questions based on a user input prompt.

```python
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
from langchain_community.llms import HuggingFacePipeline

from transformers import pipeline

# Load a free model (no API key needed)
pipe = pipeline(
    "text-generation",
    model="tiiuae/falcon-7b-instruct",   # free model
    max_new_tokens=200,
    temperature=0.7
)

llm = HuggingFacePipeline(pipeline=pipe)

# Prompt
prompt = ChatPromptTemplate.from_messages([
    ("system", "You are a helpful AI assistant. Answer clearly and concisely."),
    ("human", "{question}")
])

# Chain
chain = prompt | llm | StrOutputParser()

# Test
questions = [
    "What is machine learning?",
    "Explain supervised vs unsupervised learning simply.",
    "What is the capital of France?"
]

print("---- Batch Test ----\n")
for q in questions:
    print(f"Q: {q}")
    response = chain.invoke({"question": q})
    print(f"A: {response}\n")

# Interactive
print("\n--- Interactive Mode (type 'exit' to quit) ---")

while True:
    user_question = input("You: ")

    if user_question.lower() in ["exit", "quit"]:
        break

    answer = chain.invoke({"question": user_question})
    print(f"AI: {answer}\n")
```

## Question 8: Create a LlamaIndex implementation that indexes a local text file and retrieves answers from it.

```python
!pip install llama-index llama-index-llms-huggingface llama-index-embeddings-huggingface transformers
```

```python
# ================= INSTALL (RUN ONCE) =================
# !pip install llama-index llama-index-embeddings-huggingface transformers sentence-transformers

# ================= IMPORTS =================
from google.colab import files
from llama_index.core import VectorStoreIndex, SimpleDirectoryReader, Settings
from llama_index.core.retrievers import VectorIndexRetriever
from llama_index.core.node_parser import SentenceSplitter
from llama_index.core.ingestion import IngestionPipeline
from llama_index.embeddings.huggingface import HuggingFaceEmbedding
from llama_index.core.llms import MockLLM

# Disable LLM — only retrieval needed
Settings.llm = MockLLM()

# ================= STEP 1: UPLOAD FILE =================
print("Please upload your text file...")

uploaded          = files.upload()
uploaded_filename = list(uploaded.keys())[0]

print(f"File uploaded : {uploaded_filename}")
print(f"File size     : {len(uploaded[uploaded_filename])} bytes")

# ================= STEP 2: LOAD EMBEDDING MODEL =================
print("\nLoading embedding model... please wait.")

Settings.embed_model = HuggingFaceEmbedding(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)

# Also set chunk size here
Settings.chunk_size    = 150   # small chunks = precise answers
Settings.chunk_overlap = 20

print("Embedding model loaded successfully.")

# ================= STEP 3: LOAD DOCUMENT =================
print("\nReading uploaded document...")

documents = SimpleDirectoryReader(
    input_files=[uploaded_filename]
).load_data()

print(f"Documents loaded : {len(documents)}")

print("\nDocument Preview (first 300 chars)\n")
preview = documents[0].text[:300].strip()
for line in preview.split("\n"):
    if line.strip():
        print("  " + line.strip())
print("  ...")

# ================= STEP 4: CHUNK INTO SMALL NODES =================
print("\nSplitting document into chunks...")

splitter = SentenceSplitter(
    chunk_size=150,     # ~3-4 sentences per chunk
    chunk_overlap=20
)

pipeline = IngestionPipeline(transformations=[splitter])
nodes    = pipeline.run(documents=documents)

print(f"Total chunks created : {len(nodes)}")

# ================= STEP 5: BUILD INDEX FROM NODES =================
print("Building vector index...")

index = VectorStoreIndex(nodes)

print("Index built successfully.")

# ================= STEP 6: RETRIEVER =================
# top_k=2 now returns 2 small relevant chunks, not the whole file
retriever = VectorIndexRetriever(
    index=index,
    similarity_top_k=2
)

def get_answer(query: str) -> str:
    """Retrieve only the relevant chunks for this query."""
    nodes = retriever.retrieve(query)
    if not nodes:
        return "No relevant information found in the document."

    seen   = set()
    result = []
    for node in nodes:
        chunk = node.text.strip()
        if chunk and chunk not in seen:
            seen.add(chunk)
            result.append(chunk)

    return "\n\n".join(result)

def print_answer(answer: str):
    """Print answer with clean formatting."""
    print()
    for para in answer.split("\n"):
        para = para.strip()
        if not para:
            continue
        print("  " + para)
    print()

# ================= STEP 7: PREDEFINED QUERIES =================
queries = [
    "What is supervised learning?",
    "Explain RAG",
    "What are the types of machine learning?",
    "What is the difference between LangChain and LlamaIndex?"
]

print("\nQuerying from uploaded document\n")

for i, q in enumerate(queries, start=1):
    print(f"  Query {i} : {q}")
    print(f"  Answer :")
    answer = get_answer(q)
    print_answer(answer)

# ================= STEP 8: INTERACTIVE MODE =================
print("\nInteractive Mode")
print("Ask anything from your uploaded file.")
print("Type 'quit' to exit.\n")

while True:
    user_query = input("  Your Question : ").strip()

    if user_query.lower() in ["quit", "exit", "q"]:
        print("\nSession ended.")
        break

    if not user_query:
        print("Please enter a valid question.\n")
        continue

    answer = get_answer(user_query)
    print("  Answer :")
    print_answer(answer)

print("\nProgram completed successfully.")
```

```python
# ================= INSTALL (RUN ONCE) =================
# !pip install llama-index llama-index-embeddings-huggingface
# !pip install langchain langchain-community transformers sentence-transformers

# ================= IMPORTS =================
import os
from transformers import pipeline as hf_pipeline   # RENAMED to avoid conflict

from llama_index.core import VectorStoreIndex, SimpleDirectoryReader, Settings
from llama_index.core.retrievers import VectorIndexRetriever
from llama_index.core.node_parser import SentenceSplitter
from llama_index.core.ingestion import IngestionPipeline
from llama_index.core.llms import MockLLM
from llama_index.embeddings.huggingface import HuggingFaceEmbedding

# ================= STEP 1: Create Knowledge Base Documents =================

docs_content = {
    "company_info.txt": """
    TechCorp Solutions is a leading software company founded in 2010.
    We specialize in AI, cloud computing, and data analytics.
    Our headquarters is in Bangalore, India with offices in Mumbai and Delhi.
    We have over 5000 employees and serve clients in 30+ countries.
    Our products include CloudSuite, DataPilot, and AIAssist.
    CEO: Rajesh Kumar. CTO: Priya Sharma.
    Annual Revenue: $500 million (2024).
    """,
    "products.txt": """
    CloudSuite: Our flagship cloud management platform.
    Features: Auto-scaling, cost optimization, multi-cloud support.
    Pricing: $99/month for small teams, $499/month for enterprise.

    DataPilot: Advanced analytics and business intelligence tool.
    Features: Real-time dashboards, ML-powered insights, drag-and-drop reports.
    Pricing: $149/month per user.

    AIAssist: AI-powered productivity suite for enterprises.
    Features: Document summarization, email drafting, meeting notes.
    Pricing: $79/month per user.

    All products come with 24/7 support and a 30-day free trial.
    """,
    "hr_policy.txt": """
    TechCorp HR Policies:
    - Work Hours: 9 AM to 6 PM, Monday to Friday. Flexible work-from-home 2 days/week.
    - Leave Policy: 20 days annual leave, 10 sick leaves, 3 casual leaves per year.
    - Health Insurance: Company covers up to Rs 5 lakhs per employee and family.
    - Training Budget: Rs 50,000 per year for skill development and certifications.
    - Performance Review: Bi-annual reviews in April and October.
    - Salary Increment: Average 12% for 'Exceeds Expectations' rating.
    """
}

# Save documents locally
for filename, content in docs_content.items():
    with open(filename, "w") as f:
        f.write(content)

print("Knowledge base documents created successfully.")

# ================= STEP 2: Load Embedding Model =================

print("\nLoading embedding model... please wait.")

Settings.llm         = MockLLM()
Settings.embed_model = HuggingFaceEmbedding(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)
Settings.chunk_size    = 150
Settings.chunk_overlap = 20

print("Embedding model loaded successfully.")

# ================= STEP 3: Load and Chunk Documents =================

print("\nLoading documents...")

documents = SimpleDirectoryReader(
    input_files=list(docs_content.keys())
).load_data()

print(f"Documents loaded : {len(documents)}")

splitter        = SentenceSplitter(chunk_size=150, chunk_overlap=20)
ingest_pipeline = IngestionPipeline(transformations=[splitter])  # RENAMED variable
nodes           = ingest_pipeline.run(documents=documents)

print(f"Total chunks     : {len(nodes)}")

# ================= STEP 4: Build LlamaIndex Vector Index =================

print("\nBuilding vector index...")

index     = VectorStoreIndex(nodes)
retriever = VectorIndexRetriever(index=index, similarity_top_k=3)

print("Index built successfully.")

# ================= STEP 5: Load Free LLM (Flan-T5-Base) =================

print("\nLoading LLM (flan-t5-base) on CPU... please wait.")

qa_pipeline = hf_pipeline(          # uses renamed hf_pipeline — no conflict
    "text2text-generation",
    model="google/flan-t5-base",
    max_new_tokens=200,
    device=-1
)

print("LLM loaded successfully.")

# ================= STEP 6: Memory Setup =================

chat_history = []   # simple list memory — no langchain.memory needed

# ================= STEP 7: Retrieval Function =================

def retrieve_context(query: str) -> str:
    """Use LlamaIndex to retrieve relevant chunks for the query."""
    retrieved_nodes = retriever.retrieve(query)

    seen   = set()
    chunks = []
    for node in retrieved_nodes:
        text = node.text.strip()
        if text and text not in seen:
            seen.add(text)
            chunks.append(text)

    return "\n\n".join(chunks)

# ================= STEP 8: Chat Function =================

def chat(user_input: str) -> str:
    """Full pipeline: LlamaIndex retrieval + Flan-T5 generation + manual memory."""

    # Step 1: Retrieve relevant context
    context = retrieve_context(user_input)

    # Step 2: Build history string from last 2 exchanges
    history_text = ""
    for entry in chat_history[-4:]:
        if entry["role"] == "user":
            history_text += f"User: {entry['content']}\n"
        else:
            history_text += f"Bot: {entry['content']}\n"

    # Step 3: Build prompt
    if history_text:
        prompt = (
            f"You are a helpful assistant for TechCorp Solutions.\n"
            f"Use the context below to answer the question.\n\n"
            f"Context:\n{context}\n\n"
            f"Conversation so far:\n{history_text}\n"
            f"Question: {user_input}\n"
            f"Answer:"
        )
    else:
        prompt = (
            f"You are a helpful assistant for TechCorp Solutions.\n"
            f"Use the context below to answer the question.\n\n"
            f"Context:\n{context}\n\n"
            f"Question: {user_input}\n"
            f"Answer:"
        )

    # Step 4: Generate response
    result   = qa_pipeline(prompt)
    response = result[0]["generated_text"].strip()

    # Step 5: Fallback if model gives empty output
    if not response or len(response) < 5:
        response = context if context else "Sorry, I could not find relevant information."

    # Step 6: Save to memory
    chat_history.append({"role": "user", "content": user_input})
    chat_history.append({"role": "bot",  "content": response})

    return response

# ================= STEP 9: Demo Conversation =================

print("\nTechCorp Document Q&A Chatbot")
print("Powered by LlamaIndex + Flan-T5 + Manual Memory")
print("Type 'quit' to exit\n")

demo_questions = [
    "Who is the CEO of TechCorp?",
    "What are the features of CloudSuite and how much does it cost?",
    "How many annual leaves do employees get?",
    "Tell me more about the training budget policy."
]

for question in demo_questions:
    print(f"User : {question}")
    answer = chat(question)
    print(f"Bot  : {answer}")
    print()

# ================= STEP 10: Interactive Mode =================

print("\nNow try your own questions.\n")

while True:
    user_input = input("You : ").strip()

    if user_input.lower() in ["quit", "exit", "q"]:
        print("Goodbye!")
        break

    if not user_input:
        print("Please enter a valid question.\n")
        continue

    response = chat(user_input)
    print(f"Bot  : {response}\n")

print("\nProgram completed successfully.")
```

## Question 10: A legal firm wants to use AI to summarize large volumes of legal documents and retrieve relevant information quickly. Propose a solution using Langchain and LlamaIndex, and explain how it would work in practice.

**Proposed Solution:**

### Architecture Overview

The solution combines **LlamaIndex** (for indexing and retrieval) and **LangChain** (for summarization and orchestration) into a **Legal Intelligence Platform** with the following pipeline:

1. **Document Ingestion:** Load all legal documents (PDFs, Word files, plain text) using LlamaIndex data connectors
2. **Chunking & Indexing:** Split documents into semantic chunks and build a Vector Store Index using embeddings
3. **Retrieval (RAG):** On user query, retrieve the most relevant document chunks using semantic similarity
4. **Summarization:** Use LangChain's summarization chain (Map-Reduce) to summarize large documents
5. **Q&A:** Answer specific legal questions grounded in the retrieved documents

### How It Works in Practice
- A lawyer uploads 500 case files → system indexes all documents overnight
- Lawyer asks: "Find all cases involving breach of contract in 2022" → system retrieves and ranks relevant cases
- Lawyer asks: "Summarize this 100-page contract" → Map-Reduce chain summarizes it in seconds
- All answers are grounded in actual documents, reducing hallucination risk
