### Question 1  
**What is Detectron2 and how does it differ from previous object detection frameworks?**

**Answer:**  

Detectron2 is an open-source object detection and computer vision framework developed by Facebook AI Research (FAIR). It is built on top of PyTorch and provides state-of-the-art implementations of object detection, instance segmentation, panoptic segmentation, and keypoint detection algorithms.

**Key Characteristics of Detectron2:**
- Modular and highly configurable architecture
- Built using PyTorch for flexibility and research friendliness
- Supports advanced models like Faster R-CNN, Mask R-CNN, RetinaNet, and Cascade R-CNN
- Strong support for COCO-style datasets and evaluation metrics

**Differences from Previous Frameworks:**
- Earlier frameworks (like Detectron1 or traditional OpenCV-based methods) were less modular and harder to customize
- Detectron2 offers faster training, better GPU utilization, and cleaner APIs
- Improved support for custom datasets and research experimentation
- Easier integration with modern deep learning workflows

Overall, Detectron2 represents a significant improvement in performance, usability, and scalability over previous object detection frameworks.


### Question 2  
**Explain the process and importance of data annotation when working with
Detectron2.**

**Answer:**  

Data annotation is the process of labeling objects of interest in images so that a deep learning model can learn to recognize them. In Detectron2, annotations are typically stored in **COCO JSON format**.

**Data Annotation Process:**
1. Collect raw images relevant to the task
2. Use annotation tools such as LabelImg, CVAT, or Roboflow
3. Draw bounding boxes or segmentation masks around objects
4. Assign class labels to each object
5. Export annotations in COCO format

**Importance of Data Annotation:**
- Provides ground truth for supervised learning
- Directly affects model accuracy and generalization
- Ensures correct evaluation using metrics like mAP
- Enables Detectron2 to parse datasets correctly

High-quality, consistent annotations are critical for achieving reliable object detection performance.

### Question 4  
**What are evaluation curves in Detectron2, and how are metrics like mAP and IoU interpreted?**

**Answer:**  

Evaluation curves in Detectron2 are graphical representations used to assess the performance of object detection models.

**Intersection over Union (IoU):**
- Measures overlap between predicted and ground-truth bounding boxes
- IoU = Area of Overlap / Area of Union
- Higher IoU indicates better localization

**Mean Average Precision (mAP):**
- Primary evaluation metric in Detectron2
- Computes precision across different recall levels
- Averaged across multiple IoU thresholds and classes

**Interpretation:**
- Higher mAP → better detection accuracy
- IoU thresholds ensure both localization and classification quality
- Evaluation curves help compare models and tune hyperparameters

These metrics provide a standardized way to evaluate object detection models.

### Question 7  
**Annotate a dataset using any tool of your choice and convert the annotations to COCO format for Detectron2.**

**Answer:**  

To train a custom object detection model in Detectron2, the dataset must be annotated and converted into the **COCO (Common Objects in Context) format**, which is the standard format supported by Detectron2.

---

### Step 1: Dataset Annotation  

First, raw images are collected for the object detection task.  
An annotation tool such as **LabelImg**, **CVAT**, or **Roboflow** is used to manually label the images.

During annotation:
- Bounding boxes are drawn around objects
- Each object is assigned a class label
- Annotations are initially saved in formats such as Pascal VOC (XML) or YOLO (TXT)

---

### Step 2: COCO Format Overview  

The COCO format stores annotations in a single JSON file containing:
- **images**: image metadata (id, filename, width, height)
- **annotations**: bounding boxes, category IDs, and image IDs
- **categories**: list of object class names

This structure allows Detectron2 to efficiently load and train models on custom datasets.

---

### Step 3: Convert Annotations to COCO Format (Python Code)

Below is a simplified Python example showing how annotations can be converted into COCO JSON format:



```python
import json

coco_format = {
    "images": [],
    "annotations": [],
    "categories": [
        {"id": 1, "name": "person"},
        {"id": 2, "name": "car"}
    ]
}

# Example image entry
coco_format["images"].append({
    "id": 1,
    "file_name": "image1.jpg",
    "width": 640,
    "height": 480
})

# Example annotation entry
coco_format["annotations"].append({
    "id": 1,
    "image_id": 1,
    "category_id": 1,
    "bbox": [100, 150, 200, 300],  # [x, y, width, height]
    "area": 60000,
    "iscrowd": 0
})

# Save COCO annotation file
with open("annotations.json", "w") as f:
    json.dump(coco_format, f, indent=4)

print("COCO annotation file created successfully.")
```

### Question 8  
**Write a script to download pretrained weights and configure paths for training in Detectron2.**

**Answer:**  




```python
import os
from detectron2 import model_zoo
from detectron2.config import get_cfg
from detectron2.data.datasets import register_coco_instances

# 1. CONFIGURE YOUR PATHS
# Replace these with your actual local paths
DATASET_NAME = "my_custom_dataset"
TRAIN_JSON = "./data/annotations/train.json"
TRAIN_IMAGES = "./data/train_images"
OUTPUT_DIR = "./output/training_results"

# Create the output directory if it doesn't exist
os.makedirs(OUTPUT_DIR, exist_ok=True)

# 2. REGISTER YOUR DATASET
# This tells Detectron2 where to find your images and COCO-format labels
register_coco_instances(f"{DATASET_NAME}_train", {}, TRAIN_JSON, TRAIN_IMAGES)

# 3. CONFIGURE TRAINING
cfg = get_cfg()

# Load a base configuration file from the Model Zoo
# Example: Faster R-CNN with a ResNet-50 backbone
model_config_path = "COCO-Detection/faster_rcnn_R_50_FPN_3x.yaml"
cfg.merge_from_file(model_zoo.get_config_file(model_config_path))

# Set the training and test datasets
cfg.DATASETS.TRAIN = (f"{DATASET_NAME}_train",)
cfg.DATASETS.TEST = ()  # Leave empty if not using a validation set during training

# 4. DOWNLOAD & SET PRETRAINED WEIGHTS
# model_zoo.get_checkpoint_url() returns the URL for the weights
# Detectron2 will automatically download them to a cache directory on first run
cfg.MODEL.WEIGHTS = model_zoo.get_checkpoint_url(model_config_path)

# 5. ADJUST HYPERPARAMETERS
cfg.DATALOADER.NUM_WORKERS = 2
cfg.SOLVER.IMS_PER_BATCH = 2  # Adjust based on GPU memory
cfg.SOLVER.BASE_LR = 0.00025   # Pick a learning rate
cfg.SOLVER.MAX_ITER = 1000     # Number of training iterations
cfg.MODEL.ROI_HEADS.BATCH_SIZE_PER_IMAGE = 128
cfg.MODEL.ROI_HEADS.NUM_CLASSES = 1  # IMPORTANT: Change to your number of classes

# Set the output path
cfg.OUTPUT_DIR = OUTPUT_DIR

print(f"Configuration complete. Weights will be downloaded from: {cfg.MODEL.WEIGHTS}")
print(f"Training results will be saved to: {os.path.abspath(cfg.OUTPUT_DIR)}")
```

### Question 9  
**Show the steps and code to run inference using a trained Detectron2 model on a new image.**

**Answer:**

```python
import cv2
from detectron2.engine import DefaultPredictor
from detectron2.utils.visualizer import Visualizer
from detectron2.data import MetadataCatalog
from google.colab.patches import cv2_imshow

# 1. Setup Predictor
cfg.MODEL.WEIGHTS = model_zoo.get_checkpoint_url("COCO-Detection/faster_rcnn_R_50_FPN_3x.yaml")
cfg.MODEL.ROI_HEADS.SCORE_THRESH_TEST = 0.7
predictor = DefaultPredictor(cfg)

# 2. Load Image (ensure 'input.jpg' exists or change to your path)
im = cv2.imread("input.jpg")

# 3. Run Inference
outputs = predictor(im)

# 4. Visualize and Show
v = Visualizer(im[:, :, ::-1], metadata=MetadataCatalog.get(cfg.DATASETS.TRAIN[0]))
out = v.draw_instance_predictions(outputs["instances"].to("cpu"))
cv2_imshow(out.get_image()[:, :, ::-1])
```
