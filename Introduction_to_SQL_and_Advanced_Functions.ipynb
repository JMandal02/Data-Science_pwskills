## Question 1
Explain the fundamental differences between DDL, DML, and DQL commands in SQL. Provide one example for each type of command.

---

## Answer 1

**DDL (Data Definition Language)**
Used to create and modify the structure of the database. Changes are auto-committed.

```sql
CREATE TABLE Students (
    StudentID INT PRIMARY KEY,
    StudentName VARCHAR(100) NOT NULL
);
```

**DML (Data Manipulation Language)**
Used to insert, update, or delete data inside tables. Changes can be rolled back.

```sql
INSERT INTO Students (StudentID, StudentName)
VALUES (1, 'Alice');
```

**DQL (Data Query Language)**
Used to retrieve data from tables. Does not modify anything.

```sql
SELECT StudentName FROM Students WHERE StudentID = 1;
```

## Question 2
What is the purpose of SQL constraints? Name and describe three common types of constraints, providing a simple scenario where each would be useful.

---

## Answer 2

SQL **constraints** are rules enforced on table columns to **ensure the accuracy, validity, and integrity of the data** stored in a database. They prevent invalid data from being inserted and maintain consistency across the database.

**Purpose:**
Constraints are rules on table columns to ensure accuracy and integrity of data. They prevent invalid data from being entered.

---

**1. PRIMARY KEY**
Uniquely identifies each row. Cannot be NULL or duplicate.

```sql
CREATE TABLE Employees (
    EmployeeID INT PRIMARY KEY,
    Name VARCHAR(100)
);
```
*Scenario:* Every employee needs a unique ID. PRIMARY KEY ensures no two employees share the same ID.

---

**2. FOREIGN KEY**
Links two tables. Ensures the value exists in the referenced table.

```sql
CREATE TABLE Orders (
    OrderID INT PRIMARY KEY,
    CustomerID INT,
    FOREIGN KEY (CustomerID) REFERENCES Customers(CustomerID)
);
```
*Scenario:* An order must belong to a real customer. FOREIGN KEY prevents orders for non-existent customers.

---

**3. NOT NULL**
Ensures a column cannot be left empty.

```sql
CREATE TABLE Products (
    ProductID INT PRIMARY KEY,
    ProductName VARCHAR(100) NOT NULL,
    Price DECIMAL(10,2) NOT NULL
);
```
*Scenario:* Every product must have a name and price. NOT NULL ensures these fields are always filled.

## Question 3
Explain the difference between LIMIT and OFFSET clauses in SQL. How would you use them together to retrieve the third page of results, assuming each page has 10 records?

---

## Answer 3

**LIMIT** — Controls how many rows to return.

```sql
SELECT * FROM Products LIMIT 10;
-- Returns first 10 rows
```

**OFFSET** — Skips a number of rows before returning results.

```sql
SELECT * FROM Products LIMIT 10 OFFSET 20;
-- Skips first 20 rows, returns next 10
```

**To get Page 3 (10 records per page):**

Formula: OFFSET = (Page - 1) x Records per page = (3 - 1) x 10 = 20

```sql
SELECT * FROM Products
LIMIT 10 OFFSET 20;
```

What is a Common Table Expression (CTE) in SQL, and what are its main benefits? Provide a simple SQL example demonstrating its usage.

---

## Answer 4

**What is a CTE?**
A CTE is a temporary named result set defined using the WITH keyword. It exists only for the duration of the query.

```sql
WITH cte_name AS (
    SELECT ...
)
SELECT * FROM cte_name;
```

**Main Benefits:**
- Makes complex queries easier to read
- Can be referenced multiple times in the same query
- Cleaner alternative to nested subqueries
- Supports recursive queries

---

**Example:** Find all products priced above the average price.

```sql
WITH AvgPrice AS (
    SELECT AVG(Price) AS AveragePrice
    FROM Products
)
SELECT ProductName, Price
FROM Products, AvgPrice
WHERE Price > AveragePrice;
```

The CTE calculates the average price first, then the main query uses it to filter — keeping the logic clean and readable.

## Question 5
Describe the concept of SQL Normalization and its primary goals. Briefly explain the first three normal forms (1NF, 2NF, 3NF).

---

## Answer 5

**What is Normalization?**
Normalization is the process of organizing a database to reduce duplicate data and improve data accuracy. It involves splitting large tables into smaller related ones.

**Primary Goals:**
- Eliminate duplicate data
- Prevent update, insert, and delete anomalies
- Ensure data is stored logically

---

**1NF - First Normal Form**
Each column must have single (atomic) values. No repeating groups allowed.

Not 1NF:
| StudentID | Subjects |
|-----------|----------|
| 1 | Math, Science |

In 1NF:
| StudentID | Subject |
|-----------|---------|
| 1 | Math |
| 1 | Science |

---

**2NF - Second Normal Form**
Must be in 1NF. Every non-key column must depend on the whole primary key (no partial dependency).

Not 2NF: ProductName depends only on ProductID, not the full key (OrderID, ProductID).
Fix: Move ProductName to a separate Products table.

---

**3NF - Third Normal Form**
Must be in 2NF. No non-key column should depend on another non-key column (no transitive dependency).

Not 3NF: City depends on ZipCode, and ZipCode depends on StudentID.
Fix: Move ZipCode and City to a separate Locations table.

## Question 7
**Generate a report showing `CustomerName`, `Email`, and `TotalNumberofOrders` for each customer. Include customers who have not placed any orders (show 0). Order the results by `CustomerName`.**

---

## Answer 7

### Approach
- Use a **LEFT JOIN** between `Customers` and `Orders` so that customers with no orders are still included.
- Use `COUNT(o.OrderID)` — counting from the Orders side ensures that customers with no orders get `0` (not NULL).
- Use `GROUP BY` on the customer columns.
- Use `ORDER BY CustomerName` to sort alphabetically.

### SQL Query

```sql
SELECT
    c.CustomerName,
    c.Email,
    COUNT(o.OrderID) AS TotalNumberofOrders
FROM Customers c
LEFT JOIN Orders o ON c.CustomerID = o.CustomerID
GROUP BY c.CustomerID, c.CustomerName, c.Email
ORDER BY c.CustomerName;
```

---

### Expected Output

| CustomerName     | Email                 | TotalNumberofOrders |
|---|---|---|
| Alice Wonderland | alice@example.com     | 3 |
| Bob the Builder  | bob@example.com       | 2 |
| Charlie Chaplin  | charlie@example.com   | 1 |
| Diana Prince     | diana@example.com     | 0 |

> Diana Prince has placed no orders, so her count shows **0** due to the LEFT JOIN.

---

### Why LEFT JOIN?

| Join Type | Behaviour |
|---|---|
| `INNER JOIN` | Returns only customers who HAVE placed orders (Diana would be excluded) |
| `LEFT JOIN` | Returns ALL customers including those with no orders  |

## Question 9
**Write a SQL query that uses a CTE and a Window Function (`ROW_NUMBER()` or `RANK()`) to display `CategoryName`, `ProductName`, and `Price` for the top 2 most expensive products in each category.**

---

## Answer 9

### Approach
1. **JOIN** `Products` and `Categories` tables.
2. Use `RANK()` window function **partitioned by `CategoryName`** and **ordered by `Price` descending** to rank products within each category.
3. Wrap this in a **CTE**.
4. In the outer query, filter where `price_rank <= 2` to get only the top 2 per category.

### SQL Query

```sql
WITH RankedProducts AS (
    SELECT
        c.CategoryName,
        p.ProductName,
        p.Price,
        RANK() OVER (
            PARTITION BY c.CategoryName
            ORDER BY p.Price DESC
        ) AS price_rank
    FROM Products p
    JOIN Categories c ON p.CategoryID = c.CategoryID
)
SELECT
    CategoryName,
    ProductName,
    Price
FROM RankedProducts
WHERE price_rank <= 2
ORDER BY CategoryName, price_rank;
```

|---|---|---|

| Electronics  | Wireless Earbuds     | 150.00  |

### Task 1: Top 5 Customers by Total Amount Spent

```sql
SELECT
    CONCAT(c.first_name, ' ', c.last_name) AS CustomerName,
    c.email                                AS Email,
    SUM(p.amount)                          AS TotalAmountSpent
FROM customer c
JOIN payment p ON c.customer_id = p.customer_id
GROUP BY c.customer_id, c.first_name, c.last_name, c.email
ORDER BY TotalAmountSpent DESC
LIMIT 5;
```

**What it does:**
- Joins `customer` with `payment` on `customer_id`
- Sums all payment amounts per customer
- Orders by highest total and returns only the top 5

---

### Task 2: Top 3 Movie Categories by Rental Count

```sql
SELECT
    cat.name          AS CategoryName,
    COUNT(r.rental_id) AS RentalCount
FROM rental r
JOIN inventory i   ON r.inventory_id  = i.inventory_id
JOIN film f        ON i.film_id       = f.film_id
JOIN film_category fc ON f.film_id   = fc.film_id
JOIN category cat  ON fc.category_id = cat.category_id
GROUP BY cat.category_id, cat.name
ORDER BY RentalCount DESC
LIMIT 3;
```

**What it does:**
- Traces the path: `rental → inventory → film → film_category → category`
- Counts total rentals per category
- Returns the top 3 most rented categories

---

### Task 3: Films Available per Store and Never Rented

```sql
SELECT
    i.store_id                              AS StoreID,
    COUNT(DISTINCT i.film_id)               AS TotalFilms,
    COUNT(DISTINCT CASE
        WHEN r.rental_id IS NULL
        THEN i.film_id
    END)                                    AS NeverRentedFilms
FROM inventory i
LEFT JOIN rental r ON i.inventory_id = r.inventory_id
GROUP BY i.store_id
ORDER BY i.store_id;
```

**What it does:**
- `LEFT JOIN` keeps all inventory items, including those with no rental records
- Counts total distinct films per store
- Uses `CASE WHEN r.rental_id IS NULL` to count films that were never rented

### Task 4: Total Revenue per Month for Year 2023

```sql
SELECT
    MONTH(payment_date)     AS MonthNumber,
    MONTHNAME(payment_date) AS MonthName,
    SUM(amount)             AS TotalRevenue
FROM payment
WHERE YEAR(payment_date) = 2023
GROUP BY MONTH(payment_date), MONTHNAME(payment_date)
ORDER BY MonthNumber;
```

**What it does:**
- Filters payments made in the year 2023
- Groups by month and calculates total revenue per month
- Useful for identifying peak and slow business seasons

> **Note:** The default Sakila dataset contains data from 2005–2006. If no results appear for 2023, replace `2023` with `2005` or `2006` to test the query.

### Task 5: Customers Who Rented More Than 10 Times in the Last 6 Months

```sql
SELECT
    CONCAT(c.first_name, ' ', c.last_name) AS CustomerName,
    c.email                                AS Email,
    COUNT(r.rental_id)                     AS RentalCount
FROM customer c
JOIN rental r ON c.customer_id = r.customer_id
WHERE r.rental_date >= DATE_SUB(CURDATE(), INTERVAL 6 MONTH)
GROUP BY c.customer_id, c.first_name, c.last_name, c.email
HAVING COUNT(r.rental_id) > 10
ORDER BY RentalCount DESC;
```

**What it does:**
- Filters rentals from the **last 6 months** using `DATE_SUB(CURDATE(), INTERVAL 6 MONTH)`
- Groups by customer and counts their rentals
- `HAVING COUNT > 10` keeps only highly active customers
